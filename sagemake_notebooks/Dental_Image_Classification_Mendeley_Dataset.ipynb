```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, Input
from tensorflow.keras.applications import MobileNetV2 # Using MobileNetV2 as an example base model
from tensorflow.keras.optimizers import Adam

import sagemaker
import boto3
from urllib.parse import urlparse
import os
import shutil
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, accuracy_score
from sklearn.preprocessing import label_binarize
from itertools import cycle
import glob # To list files easily

print(f"TensorFlow Version: {tf.__version__}")
print(f"SageMaker SDK Version: {sagemaker.__version__}")

# --- 1. Configuration ---
s3_paths = {
    "Broken Teeth": "s3://sagemaker-studio-010526272250-hsf94lgtf6i/Mendeley_Data/Dental OPG (Classification)/BDC-BDR/",
    "Caries": "s3://sagemaker-studio-010526272250-hsf94lgtf6i/Mendeley_Data/Dental OPG (Classification)/Caries/",
    "Fractured Teeth": "s3://sagemaker-studio-010526272250-hsf94lgtf6i/Mendeley_Data/Dental OPG (Classification)/Fractured Teeth/",
    "Healthy Teeth": "s3://sagemaker-studio-010526272250-hsf94lgtf6i/Mendeley_Data/Dental OPG (Classification)/Healthy Teeth/",
    "Impacted Teeth": "s3://sagemaker-studio-010526272250-hsf94lgtf6i/Mendeley_Data/Dental OPG (Classification)/Impacted teeth/", # Corrected key name
    "Infected Teeth": "s3://sagemaker-studio-010526272250-hsf94lgtf6i/Mendeley_Data/Dental OPG (Classification)/Infection/"   # Corrected key name
}

# Make class names filesystem-friendly (replace spaces)
class_names = [name.replace(" ", "_") for name in s3_paths.keys()]
s3_paths_clean_keys = {name.replace(" ", "_"): path for name, path in s3_paths.items()}
num_classes = len(class_names)

print(f"Classes found: {class_names}")
print(f"Number of classes: {num_classes}")

# Local directories within the SageMaker Studio environment
base_local_dir = '/tmp/dental_data' # Use /tmp as it's usually faster for temporary data
raw_data_dir = os.path.join(base_local_dir, 'raw')
processed_data_dir = os.path.join(base_local_dir, 'processed')
train_dir = os.path.join(processed_data_dir, 'train')
validation_dir = os.path.join(processed_data_dir, 'validation')
test_dir = os.path.join(processed_data_dir, 'test')

# Model/Training Parameters
IMG_HEIGHT = 224
IMG_WIDTH = 224
IMG_SHAPE = (IMG_HEIGHT, IMG_WIDTH, 3)
BATCH_SIZE = 32
EPOCHS = 25 # Adjust as needed based on convergence
LEARNING_RATE = 0.0001
VAL_SPLIT = 0.15 # Proportion of data for validation
TEST_SPLIT = 0.15 # Proportion of data for testing

# --- 2. Data Download and Preparation ---

def download_s3_folder(s3_uri, local_dir):
    """Downloads files from an S3 prefix to a local directory."""
    if os.path.exists(local_dir):
        print(f"Directory {local_dir} already exists. Skipping download for this folder.")
        # Optional: Add logic here to check if content is complete or redownload if needed
        # For simplicity, we assume existing directory means data is present.
        # Be cautious: if a previous run failed, the data might be incomplete.
        # Consider removing the directory first for a clean download:
        # print(f"Removing existing directory: {local_dir}")
        # shutil.rmtree(local_dir)
    else:
         os.makedirs(local_dir, exist_ok=True)
         print(f"Downloading data from {s3_uri} to {local_dir}...")
         try:
             sagemaker.s3.S3Downloader.download(s3_uri, local_dir)
             print(f"Successfully downloaded data from {s3_uri}")
         except Exception as e:
             print(f"Error downloading {s3_uri}: {e}")
             # Handle error appropriately, maybe skip this class or raise exception

# Download data for each class
print("\n--- Starting Data Download ---")
for class_name, s3_path in s3_paths_clean_keys.items():
    local_class_dir = os.path.join(raw_data_dir, class_name)
    download_s3_folder(s3_path, local_class_dir)

print("\n--- Data Download Phase Complete ---")


def split_data(source_dir, train_dir, validation_dir, test_dir, classes, val_split, test_split):
    """Splits data from source folders into train, validation, and test sets."""
    print(f"\n--- Starting Data Splitting ---")
    # Ensure target directories exist and are empty
    for dir_path in [train_dir, validation_dir, test_dir]:
        if os.path.exists(dir_path):
            print(f"Removing existing directory: {dir_path}")
            shutil.rmtree(dir_path)
        os.makedirs(dir_path)
        for cls in classes:
            os.makedirs(os.path.join(dir_path, cls), exist_ok=True)

    total_files_processed = 0
    for cls in classes:
        class_source_dir = os.path.join(source_dir, cls)
        if not os.path.isdir(class_source_dir):
             print(f"Warning: Source directory for class '{cls}' not found at {class_source_dir}. Skipping.")
             continue

        # Use glob to find common image types (adjust extensions if needed)
        all_files = glob.glob(os.path.join(class_source_dir, '*.png')) + \
                    glob.glob(os.path.join(class_source_dir, '*.jpg')) + \
                    glob.glob(os.path.join(class_source_dir, '*.jpeg')) + \
                    glob.glob(os.path.join(class_source_dir, '*.bmp')) + \
                    glob.glob(os.path.join(class_source_dir, '*.tif'))

        if not all_files:
             print(f"Warning: No image files found in {class_source_dir} for class '{cls}'. Skipping.")
             continue

        n_files = len(all_files)
        total_files_processed += n_files
        print(f"Found {n_files} images for class: {cls}")

        # First split: Separate out test set
        train_val_files, test_files = train_test_split(all_files, test_size=test_split, random_state=42)

        # Second split: Split remaining into train and validation
        # Adjust validation split percentage based on the remaining data
        relative_val_split = val_split / (1.0 - test_split)
        if relative_val_split >= 1.0: # Handle cases where splits are too large
             print(f"Warning: Validation split percentage is too high for class {cls}. Adjusting.")
             # Assign all remaining to train, or handle as error? For now, assign rest to train.
             train_files = train_val_files
             val_files = []
        else:
            train_files, val_files = train_test_split(train_val_files, test_size=relative_val_split, random_state=42)

        # Copy files to respective directories
        def copy_files(files, dest_folder):
            for f in files:
                try:
                    shutil.copy(f, os.path.join(dest_folder, cls))
                except Exception as e:
                    print(f"Error copying file {f} to {dest_folder}/{cls}: {e}")

        copy_files(train_files, train_dir)
        copy_files(val_files, validation_dir)
        copy_files(test_files, test_dir)

        print(f"Class '{cls}': {len(train_files)} train, {len(val_files)} validation, {len(test_files)} test")

    if total_files_processed == 0:
        raise ValueError("No image files were found in any of the raw data directories. Please check S3 paths and downloaded content.")

    print(f"\n--- Data Splitting Complete ---")

# Split the downloaded raw data
split_data(raw_data_dir, train_dir, validation_dir, test_dir, class_names, VAL_SPLIT, TEST_SPLIT)

# --- 3. Data Augmentation and Generators ---
print("\n--- Setting up Data Generators ---")

# Augmentation for training data
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.15,
    height_shift_range=0.15,
    shear_range=0.15,
    zoom_range=0.15,
    horizontal_flip=True,
    fill_mode='nearest'
    # brightness_range=[0.8, 1.2] # Optional: Add brightness augmentation
    )

# Only rescaling for validation and test data (NO augmentation)
validation_test_datagen = ImageDataGenerator(rescale=1./255)

# Flow data from directories
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical', # for multi-class classification
    classes=class_names,      # Ensure order is consistent
    shuffle=True)

validation_generator = validation_test_datagen.flow_from_directory(
    validation_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    classes=class_names,
    shuffle=False) # No need to shuffle validation data

test_generator = validation_test_datagen.flow_from_directory(
    test_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=1,  # Process one image at a time for evaluation
    class_mode='categorical',
    classes=class_names,
    shuffle=False) # Do not shuffle test data

# Verify class indices
print(f"Class Indices from Generators: {train_generator.class_indices}")
if train_generator.num_classes != num_classes:
     print(f"Warning: Number of classes found by generator ({train_generator.num_classes}) doesn't match expected ({num_classes}). Check subdirectories.")

# --- 4. Model Building (Transfer Learning with MobileNetV2) ---
print("\n--- Building Model ---")

# Load the pre-trained base model without the classification head
base_model = MobileNetV2(input_shape=IMG_SHAPE,
                         include_top=False, # Exclude the final Dense layer
                         weights='imagenet')

# Freeze the base model layers
base_model.trainable = False

# Create the new model on top
inputs = Input(shape=IMG_SHAPE)
x = base_model(inputs, training=False) # Run base model in inference mode (for frozen layers)
x = GlobalAveragePooling2D()(x)
x = Dropout(0.3)(x) # Add dropout for regularization
x = Dense(128, activation='relu')(x) # Add a dense layer
outputs = Dense(num_classes, activation='softmax')(x) # Final layer with softmax for multi-class

model = Model(inputs, outputs)

# Compile the model
model.compile(optimizer=Adam(learning_rate=LEARNING_RATE),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()

# --- 5. Model Training ---
print("\n--- Starting Model Training ---")

# Add callbacks (optional but recommended)
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
# model_checkpoint = tf.keras.callbacks.ModelCheckpoint('best_dental_model.h5', save_best_only=True, monitor='val_loss')

# Calculate steps per epoch
steps_per_epoch = train_generator.samples // BATCH_SIZE
validation_steps = validation_generator.samples // BATCH_SIZE
# Adjust if generators might not yield full batches at the end
if train_generator.samples % BATCH_SIZE > 0:
    steps_per_epoch += 1
if validation_generator.samples % BATCH_SIZE > 0:
    validation_steps += 1


history = model.fit(
    train_generator,
    steps_per_epoch=steps_per_epoch,
    epochs=EPOCHS,
    validation_data=validation_generator,
    validation_steps=validation_steps,
    callbacks=[early_stopping] # Add model_checkpoint here if desired
)

print("\n--- Model Training Complete ---")

# --- 6. Evaluation ---
print("\n--- Evaluating Model ---")

# A. Training and Validation Accuracy from History
train_acc = history.history['accuracy'][-1]
val_acc = history.history['val_accuracy'][-1]
print(f"Final Training Accuracy: {train_acc:.4f}")
print(f"Final Validation Accuracy: {val_acc:.4f}")

# Plot Training history
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.tight_layout()
plt.show()


# B. Test Set Evaluation
print("\nEvaluating on Test Set...")
test_loss, test_acc = model.evaluate(test_generator, steps=test_generator.samples // test_generator.batch_size)
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_acc:.4f}")

# C. Predictions and Detailed Metrics
print("\nGenerating Predictions for Detailed Metrics...")
# Reset the test generator before predicting to ensure order
test_generator.reset()
Y_pred_probs = model.predict(test_generator, steps=test_generator.samples // test_generator.batch_size)
# Get the index of the highest probability class
y_pred_indices = np.argmax(Y_pred_probs, axis=1)

# Get true labels
y_true_indices = test_generator.classes # Relies on shuffle=False in test_generator

# Ensure we have the same number of predictions and true labels
if len(y_pred_indices) != len(y_true_indices):
     print(f"Warning: Number of predictions ({len(y_pred_indices)}) does not match number of true labels ({len(y_true_indices)}). Evaluation might be inaccurate.")
     # Possible fix: Adjust steps in predict/evaluate if dataset size isn't divisible by batch size
     # For batch_size=1, steps should ideally be test_generator.samples
     test_generator.reset()
     Y_pred_probs = model.predict(test_generator, steps=test_generator.samples)
     y_pred_indices = np.argmax(Y_pred_probs, axis=1)
     y_true_indices = test_generator.classes[:len(y_pred_indices)] # Ensure matching length if prediction stopped early


# Get class names in the order the generator used them
generator_class_names = list(test_generator.class_indices.keys())
print(f"Classes used for report/matrix: {generator_class_names}")

# Classification Report
print("\nClassification Report:")
print(classification_report(y_true_indices, y_pred_indices, target_names=generator_class_names))

# Confusion Matrix
print("\nConfusion Matrix:")
cm = confusion_matrix(y_true_indices, y_pred_indices)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=generator_class_names, yticklabels=generator_class_names)
plt.title('Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()


# D. ROC Curves (One-vs-Rest)
print("\nGenerating ROC Curves...")

# Binarize the true labels
y_true_bin = label_binarize(y_true_indices, classes=range(num_classes))

# Calculate ROC curve and AUC for each class
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(num_classes):
    fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], Y_pred_probs[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plot all ROC curves
plt.figure(figsize=(10, 8))
colors = cycle(['aqua', 'darkorange', 'cornflowerblue', 'green', 'red', 'purple', 'brown']) # Add more colors if needed

for i, color in zip(range(num_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2,
             label=f'ROC curve of class {generator_class_names[i]} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--', lw=2) # Plot diagonal line
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) - One-vs-Rest')
plt.legend(loc="lower right")
plt.show()

print("\n--- Evaluation Complete ---")

# --- 7. Optional: Cleanup ---
# print("\nCleaning up downloaded data...")
# try:
#     shutil.rmtree(base_local_dir)
#     print(f"Successfully removed {base_local_dir}")
# except Exception as e:
#     print(f"Error removing temporary data directory {base_local_dir}: {e}")
```

```python
# ******************************************************************************
# ** CRITICAL SETUP STEP FOR SAGEMAKER **
# ******************************************************************************
# If you are seeing the error:
# "ImportError: libGL.so.1: cannot open shared object file: No such file or directory"
# This means a system library needed by OpenCV (cv2) is missing.
#
# YOU MUST FIX THIS MANUALLY *BEFORE* RUNNING THIS PYTHON CODE:
# 1. Open a Terminal in SageMaker (File -> New -> Terminal).
# 2. Run this EXACT command in the terminal:
#    sudo yum install -y mesa-libGL
# 3. Wait for the installation to complete.
# 4. Go back to this notebook.
# 5. Restart the Kernel (Kernel -> Restart Kernel).
# 6. NOW, you can run this Python code cell.
# ******************************************************************************

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, Input
from tensorflow.keras.applications import MobileNetV2 # Or another suitable base model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping # ModelCheckpoint (optional)

# Attempt to import OpenCV AFTER potentially running the fix
try:
    import cv2 # Using OpenCV for image manipulation
    print(f"Successfully imported OpenCV. Version: {cv2.__version__}")
except ImportError as e:
    print(f"FATAL ERROR importing OpenCV: {e}")
    print("This usually means the 'libGL.so.1' library is still missing.")
    print("PLEASE FOLLOW THE 'CRITICAL SETUP STEP' instructions at the top of this script.")
    print("1. Open Terminal")
    print("2. Run: sudo yum install -y mesa-libGL")
    print("3. Restart Kernel")
    print("4. Re-run this code cell.")
    # Stop execution if cv2 cannot be imported, as it's essential
    raise e # Re-raise the error to halt the script

import sagemaker
import boto3
from urllib.parse import urlparse
import os
import shutil
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, accuracy_score
from sklearn.preprocessing import label_binarize
from sklearn.utils import class_weight # Import for class weights
from itertools import cycle
import glob # To list files easily
import time # For timing download/processing

print(f"TensorFlow Version: {tf.__version__}")
print(f"SageMaker SDK Version: {sagemaker.__version__}")

# --- 1. Configuration ---
s3_paths = {
    "Broken Teeth": "s3://sagemaker-studio-010526272250-hsf94lgtf6i/Mendeley_Data/Dental OPG (Classification)/BDC-BDR/",
    "Caries": "s3://sagemaker-studio-010526272250-hsf94lgtf6i/Mendeley_Data/Dental OPG (Classification)/Caries/",
    "Fractured Teeth": "s3://sagemaker-studio-010526272250-hsf94lgtf6i/Mendeley_Data/Dental OPG (Classification)/Fractured Teeth/",
    "Healthy Teeth": "s3://sagemaker-studio-010526272250-hsf94lgtf6i/Mendeley_Data/Dental OPG (Classification)/Healthy Teeth/",
    "Impacted Teeth": "s3://sagemaker-studio-010526272250-hsf94lgtf6i/Mendeley_Data/Dental OPG (Classification)/Impacted teeth/",
    "Infected Teeth": "s3://sagemaker-studio-010526272250-hsf94lgtf6i/Mendeley_Data/Dental OPG (Classification)/Infection/"
}

# Make class names filesystem-friendly (replace spaces) and sort
class_names = sorted([name.replace(" ", "_") for name in s3_paths.keys()])
s3_paths_clean_keys = {name.replace(" ", "_"): path for name, path in s3_paths.items()}
num_classes = len(class_names)

print(f"Classes found: {class_names}")
print(f"Number of classes: {num_classes}")

# Local directories
base_local_dir = '/tmp/dental_data_quadrants' # New base dir for quadrant data
raw_data_dir = os.path.join(base_local_dir, 'raw')
processed_data_dir = os.path.join(base_local_dir, 'processed') # This will hold quadrant images
train_dir = os.path.join(processed_data_dir, 'train')
validation_dir = os.path.join(processed_data_dir, 'validation')
test_dir = os.path.join(processed_data_dir, 'test')

# Model/Training Parameters --- ADJUSTED FOR QUADRANTS ---
QUAD_IMG_HEIGHT = 112
QUAD_IMG_WIDTH = 112
# Important: Ensure input images are read in color (3 channels) for MobileNetV2
QUAD_IMG_SHAPE = (QUAD_IMG_HEIGHT, QUAD_IMG_WIDTH, 3)

BATCH_SIZE = 32
EPOCHS = 50 # Rely on EarlyStopping
INITIAL_LEARNING_RATE = 1e-4
FINE_TUNE_LEARNING_RATE = 1e-5
VAL_SPLIT = 0.15
TEST_SPLIT = 0.15
RANDOM_STATE = 42 # For reproducible splits
DROPOUT_RATE = 0.4
DENSE_UNITS = 256
FINE_TUNE_AT_LAYER = 100 # Unfreeze layers from this index onwards for fine-tuning
EARLY_STOP_PATIENCE_STAGE1 = 10
EARLY_STOP_PATIENCE_STAGE2 = 10 # Can be same or different

# --- 2. Data Download ---

def download_s3_folder(s3_uri, local_dir):
    """Downloads files from an S3 prefix to a local directory."""
    if not os.path.exists(local_dir):
        os.makedirs(local_dir, exist_ok=True)
        print(f"Created directory: {local_dir}")
    else:
        print(f"Directory {local_dir} already exists. Skipping creation.")
        # Optional: Clean directory before download if needed
        # print(f"Cleaning existing directory: {local_dir}")
        # shutil.rmtree(local_dir)
        # os.makedirs(local_dir, exist_ok=True)

    print(f"Downloading data from {s3_uri} to {local_dir}...")
    start_time = time.time()
    try:
        sagemaker.s3.S3Downloader.download(s3_uri, local_dir)
        elapsed_time = time.time() - start_time
        # Check if files were actually downloaded (simple check)
        if not os.listdir(local_dir):
             print(f"Warning: Download from {s3_uri} completed, but directory {local_dir} is empty. Check S3 path and permissions.")
        else:
             print(f"Successfully downloaded data from {s3_uri} in {elapsed_time:.2f} seconds")
    except Exception as e:
        print(f"Error downloading {s3_uri}: {e}")
        # Decide if the script should stop if download fails
        raise e # Reraise to stop execution

print("\n--- Starting Data Download ---")
# Clean raw_data_dir before downloading to ensure freshness
if os.path.exists(raw_data_dir):
    print(f"Removing existing raw data directory: {raw_data_dir}")
    shutil.rmtree(raw_data_dir)
os.makedirs(raw_data_dir) # Recreate after cleaning

download_success = True
for class_name in class_names:
    s3_path = s3_paths_clean_keys.get(class_name)
    if s3_path:
        local_class_dir = os.path.join(raw_data_dir, class_name)
        try:
            download_s3_folder(s3_path, local_class_dir)
        except Exception as e:
            print(f"Download failed for class '{class_name}', stopping execution.")
            download_success = False
            break # Stop if one class fails
    else:
        print(f"Warning: Could not find S3 path for class '{class_name}'.")
        # Consider if this is critical - maybe set download_success = False

if not download_success:
    print("\n--- Data Download Failed. Exiting. ---")
    exit()
else:
    print("\n--- Data Download Phase Complete ---")


# --- 3. Data Splitting and Quadrant Generation ---

def split_and_generate_quadrants(source_dir, train_dir, validation_dir, test_dir, classes, val_split, test_split, target_quad_size=(QUAD_IMG_WIDTH, QUAD_IMG_HEIGHT), random_state=RANDOM_STATE):
    """
    Splits original image file paths and generates quadrant images,
    saving them to the respective train/val/test directories.
    ASSUMES: The label of the original image applies to all its quadrants.
    """
    print(f"\n--- Starting Data Splitting and Quadrant Generation ---")
    print(f"!!! WARNING: Assigning the original image label to ALL its quadrants. !!!")
    print(f"!!! This assumes the condition is representative across the image.   !!!")

    # Ensure target directories exist and are empty
    for dir_path in [processed_data_dir, train_dir, validation_dir, test_dir]: # Include processed_data_dir for clean start
        if os.path.exists(dir_path):
            print(f"Removing existing directory: {dir_path}")
            shutil.rmtree(dir_path)
    # Recreate necessary directories
    os.makedirs(processed_data_dir)
    os.makedirs(train_dir)
    os.makedirs(validation_dir)
    os.makedirs(test_dir)
    for cls in classes:
        os.makedirs(os.path.join(train_dir, cls), exist_ok=True)
        os.makedirs(os.path.join(validation_dir, cls), exist_ok=True)
        os.makedirs(os.path.join(test_dir, cls), exist_ok=True)

    total_original_files = 0
    total_quadrants_generated = 0
    split_summary = {} # Store counts per class

    # Dictionary to store original file paths for each set (needed for later inference example)
    original_test_files_map = {cls: [] for cls in classes}

    for cls in classes:
        class_source_dir = os.path.join(source_dir, cls)
        if not os.path.isdir(class_source_dir):
             print(f"Warning: Source directory for class '{cls}' not found at {class_source_dir}. Skipping.")
             continue

        # Find original image files (more robustly handles extensions)
        supported_extensions = ['*.png', '*.jpg', '*.jpeg', '*.bmp', '*.tif', '*.tiff']
        all_files = []
        for ext in supported_extensions:
            # Use case-insensitive globbing if possible (though less standard) or list both lower/upper
            all_files.extend(glob.glob(os.path.join(class_source_dir, ext)))
            all_files.extend(glob.glob(os.path.join(class_source_dir, ext.upper())))

        # Remove duplicates that might arise from case differences
        all_files = sorted(list(set(all_files)))


        if not all_files:
             print(f"Warning: No supported original image files found in {class_source_dir} for class '{cls}'. Skipping.")
             continue

        n_files = len(all_files)
        total_original_files += n_files
        print(f"\nFound {n_files} original images for class: {cls}")

        # Stratified Split Logic (more robust)
        train_files_orig, test_files_orig = [], []
        val_files_orig = []

        # Need at least 1 sample per set for stratified split if splits > 0 and < 1
        min_samples_needed = 1
        if 0 < test_split < 1: min_samples_needed += 1
        if 0 < val_split < 1: min_samples_needed += 1


        if n_files < min_samples_needed:
            print(f"Warning: Class '{cls}' has only {n_files} samples. Cannot perform specified stratified split with min {min_samples_needed} samples. Assigning all to training.")
            train_files_orig = all_files
        else:
            try:
                # Split test set first
                actual_test_size = 0.0 # Initialize
                if test_split > 0:
                    # Ensure test_size results in at least 1 sample remaining for train/val combined if val_split > 0
                    min_remaining_for_tv = 1 if val_split == 0 else 2 # Need 2 if we might split val later
                    max_test_count = n_files - min_remaining_for_tv
                    if max_test_count <= 0:
                         print(f"Warning: Not enough samples ({n_files}) in class '{cls}' to split off test data while preserving samples for train/validation based on splits. Setting test split to 0 for this class.")
                         actual_test_size = 0.0
                         train_val_files = all_files
                    else:
                        actual_test_size = min(test_split, max_test_count / n_files) # Adjust test_split if needed
                        actual_test_size = max(0, actual_test_size) # Ensure non-negative

                        if actual_test_size == 0 and test_split > 0:
                             print(f"Warning: Calculated test split is 0 for class '{cls}' due to sample constraints. No test samples will be split.")
                             train_val_files = all_files
                        elif n_files * actual_test_size < 1:
                             print(f"Warning: Calculated test split ({actual_test_size}) results in < 1 sample for class '{cls}'. Setting test split to 0.")
                             actual_test_size = 0.0
                             train_val_files = all_files
                        else:
                            # Check sklearn condition for stratification: n_splits <= n_samples_per_class
                            # Here n_splits is implicitly 2 (test vs rest)
                            # We need at least 2 samples to stratify
                            if n_files < 2: # Should have been caught by min_samples_needed, but belt-and-suspenders
                                 print(f"Warning: Only {n_files} sample(s) in class '{cls}', cannot stratify test split. Using non-stratified.")
                                 train_val_files, test_files_orig = train_test_split(all_files, test_size=actual_test_size, random_state=random_state)
                            else:
                                train_val_files, test_files_orig = train_test_split(
                                    all_files,
                                    test_size=actual_test_size,
                                    random_state=random_state,
                                    stratify=[cls] * n_files # Use dummy labels for stratification
                                )
                else: # test_split is 0
                    train_val_files = all_files

                # Split validation set from remaining train_val
                actual_val_size = 0.0 # Initialize
                if val_split > 0 and len(train_val_files) > 1: # Need at least 2 samples to split validation from train_val
                    # Calculate relative validation split size from the train_val pool
                    denominator = (1.0 - actual_test_size) if (1.0 - actual_test_size) > 0 else 1.0
                    relative_val_split = val_split / denominator
                    # Ensure val split leaves at least 1 sample for training
                    max_val_count = len(train_val_files) - 1
                    if max_val_count <= 0:
                         print(f"Warning: Not enough samples remaining ({len(train_val_files)}) in class '{cls}' to split off validation. Assigning all remaining to train.")
                         actual_val_size = 0.0
                         train_files_orig = train_val_files
                    else:
                        actual_val_size = min(relative_val_split, max_val_count / len(train_val_files))
                        actual_val_size = max(0, actual_val_size)

                        if actual_val_size == 0 and val_split > 0:
                            print(f"Warning: Calculated validation split is 0 for class '{cls}' due to sample constraints. No validation samples will be split.")
                            train_files_orig = train_val_files
                        elif len(train_val_files) * actual_val_size < 1:
                            print(f"Warning: Calculated validation split ({actual_val_size}) results in < 1 sample for class '{cls}'. Setting validation split to 0.")
                            actual_val_size = 0.0
                            train_files_orig = train_val_files
                        else:
                            # Check sklearn condition for stratification (need >= 2 samples)
                            if len(train_val_files) < 2:
                                print(f"Warning: Only {len(train_val_files)} sample(s) left in class '{cls}', cannot stratify validation split. Using non-stratified.")
                                train_files_orig, val_files_orig = train_test_split(train_val_files, test_size=actual_val_size, random_state=random_state)
                            else:
                                train_files_orig, val_files_orig = train_test_split(
                                    train_val_files,
                                    test_size=actual_val_size,
                                    random_state=random_state,
                                    stratify=[cls] * len(train_val_files) # Stratify again
                                )
                else: # No validation split needed or not enough samples left
                    train_files_orig = train_val_files

            except ValueError as e:
                 print(f"Error during stratified split for class '{cls}' (n={n_files}): {e}. Performing non-stratified split as fallback.")
                 # Fallback to non-stratified (less ideal)
                 train_val_files, test_files_orig = train_test_split(all_files, test_size=test_split, random_state=random_state)
                 if val_split > 0 and len(train_val_files) > 0:
                     relative_val_split = val_split / (1.0 - test_split) if (1.0 - test_split) > 0 else 0
                     if relative_val_split < 1.0 and len(train_val_files) > 1 :
                          train_files_orig, val_files_orig = train_test_split(train_val_files, test_size=relative_val_split, random_state=random_state)
                     else:
                          train_files_orig = train_val_files
                 else:
                    train_files_orig = train_val_files


        print(f"Originals split for '{cls}': {len(train_files_orig)} train, {len(val_files_orig)} validation, {len(test_files_orig)} test")

        # Store original test file paths
        original_test_files_map[cls] = test_files_orig

        # --- Process and Save Quadrants ---
        def process_and_save(original_files, dest_dir, set_name):
            quad_count = 0
            processed_basenames = set() # Track base filenames to avoid duplicates if glob pattern matches variants like .png and .PNG
            for f_orig in original_files:
                # Normalize path separators and get basename without extension
                try:
                    norm_path = os.path.normpath(f_orig)
                    base_filename = os.path.splitext(os.path.basename(norm_path))[0]
                except Exception as path_e:
                     print(f"Warning: Could not process filename {f_orig}: {path_e}. Skipping.")
                     continue

                if base_filename in processed_basenames:
                    # print(f"Debug: Skipping duplicate base filename {base_filename} from path {f_orig}")
                    continue # Skip if already processed (e.g., .png and .PNG version)
                processed_basenames.add(base_filename)

                try:
                    # Read image - Ensure it's read in color
                    img = cv2.imread(f_orig)
                    if img is None:
                        print(f"Warning: Could not read image {f_orig} using OpenCV. Skipping.")
                        continue

                    # Ensure 3 channels (BGR for OpenCV)
                    if len(img.shape) == 2 or img.shape[2] == 1: # Grayscale
                        img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
                    elif img.shape[2] == 4: # Handle BGRA/RGBA
                        img = cv2.cvtColor(img, cv2.COLOR_BGRA2BGR)
                    elif img.shape[2] != 3: # Unexpected number of channels
                         print(f"Warning: Image {f_orig} has unexpected shape {img.shape}. Skipping.")
                         continue

                    h, w, _ = img.shape
                    # Skip tiny images that cannot be meaningfully split
                    if h < 2 or w < 2:
                        print(f"Warning: Image {f_orig} is too small ({w}x{h}). Skipping quadrant generation.")
                        continue

                    h_mid, w_mid = h // 2, w // 2

                    # Define quadrants (TopLeft, TopRight, BottomLeft, BottomRight)
                    quadrants = {
                        "q1_tl": img[0:h_mid, 0:w_mid],
                        "q2_tr": img[0:h_mid, w_mid:w],
                        "q3_bl": img[h_mid:h, 0:w_mid],
                        "q4_br": img[h_mid:h, w_mid:w]
                    }

                    quads_saved_from_orig = 0
                    for q_name, q_img in quadrants.items():
                         if q_img.size == 0: # Check for empty quadrants if original dims were odd/small
                             #print(f"Debug: Quadrant {q_name} from {f_orig} is empty (orig shape {h}x{w}). Skipping.")
                             continue

                         # Resize quadrant to target size
                         # Use INTER_AREA for shrinking, INTER_LINEAR or INTER_CUBIC for enlarging (though shrinking is expected here)
                         if q_img.shape[0] > target_quad_size[1] or q_img.shape[1] > target_quad_size[0]:
                             interpolation = cv2.INTER_AREA
                         else:
                             interpolation = cv2.INTER_LINEAR # Use linear if not shrinking
                         q_img_resized = cv2.resize(q_img, target_quad_size, interpolation=interpolation)

                         # Construct save path
                         # Sanitize base_filename further if necessary (e.g., remove special chars)
                         safe_base_filename = "".join(c if c.isalnum() or c in ('-', '_') else '_' for c in base_filename)
                         quad_filename = f"{safe_base_filename}_{q_name}.png" # Standardize output format
                         save_path = os.path.join(dest_dir, cls, quad_filename)

                         # Ensure the class directory exists (it should, but double-check)
                         os.makedirs(os.path.dirname(save_path), exist_ok=True)

                         # Save the quadrant image
                         save_success = cv2.imwrite(save_path, q_img_resized)
                         if not save_success:
                             print(f"Warning: Failed to save quadrant {save_path}")
                         else:
                             quads_saved_from_orig += 1

                    # Only increment total quad_count if at least one quadrant was saved for this original image
                    if quads_saved_from_orig > 0:
                         quad_count += quads_saved_from_orig # Add number actually saved

                except cv2.error as cv_err:
                    print(f"OpenCV error processing file {f_orig}: {cv_err}. Skipping.")
                except Exception as e:
                    print(f"Error processing file {f_orig} into quadrants: {e.__class__.__name__}: {e}. Skipping.")
            return quad_count

        print(f"Generating train quadrants for class '{cls}'...")
        q_train = process_and_save(train_files_orig, train_dir, "train")
        print(f"Generating validation quadrants for class '{cls}'...")
        q_val = process_and_save(val_files_orig, validation_dir, "validation")
        print(f"Generating test quadrants for class '{cls}'...")
        q_test = process_and_save(test_files_orig, test_dir, "test")

        split_summary[cls] = {'train': q_train, 'val': q_val, 'test': q_test}
        total_quadrants_generated += (q_train + q_val + q_test)
        print(f"Generated Quadrants for '{cls}': {q_train} train, {q_val} validation, {q_test} test")


    if total_original_files == 0:
        print("CRITICAL ERROR: No original image files were found in any raw data directories. Check S3 paths and download success.")
        raise ValueError("No original images found.")
    if total_quadrants_generated == 0:
        print("CRITICAL ERROR: No quadrant images were generated. Check processing logic, source files, permissions, and image reading capabilities.")
        raise ValueError("Quadrant generation failed.")

    print("\n--- Quadrant Generation Summary ---")
    for cls, counts in split_summary.items():
        print(f"Class '{cls}': Train={counts['train']}, Val={counts['val']}, Test={counts['test']}")
    print(f"Total Original Files Found: {total_original_files}")
    print(f"Total Quadrants Generated: {total_quadrants_generated}")
    print(f"--- Quadrant Generation Complete ---")

    # Return the map of original test files for later use
    return original_test_files_map

# Run the splitting and quadrant generation
try:
    original_test_files = split_and_generate_quadrants(raw_data_dir, train_dir, validation_dir, test_dir, class_names, VAL_SPLIT, TEST_SPLIT)
except ValueError as e:
    print(f"Stopping script due to error during data preparation: {e}")
    exit(1) # Use non-zero exit code for errors
except Exception as e:
     print(f"An unexpected error occurred during data preparation: {e}")
     exit(1)


# --- 4. Calculate Class Weights (Based on *Quadrant* counts) ---
print("\n--- Calculating Class Weights for Training Quadrants ---")
class_weights = None # Initialize
try:
    # Count quadrant samples per class in the training directory more directly
    train_samples_count = {}
    all_train_labels = []
    class_label_map = {cls: i for i, cls in enumerate(class_names)} # Consistent mapping
    train_files_found = False

    print(f"Scanning training directory: {train_dir}")
    for cls, label_index in class_label_map.items():
        class_path = os.path.join(train_dir, cls)
        count = 0
        if os.path.isdir(class_path):
            # Count files robustly (handle potential hidden files, though unlikely here)
            try:
                files_in_class = [f for f in os.listdir(class_path) if os.path.isfile(os.path.join(class_path, f))]
                count = len(files_in_class)
            except OSError as list_err:
                print(f"Warning: Could not list files in {class_path}: {list_err}. Assuming 0 samples for {cls}.")
                count = 0

            if count > 0:
                train_files_found = True
                all_train_labels.extend([label_index] * count) # Append the integer label 'count' times
        else:
             print(f"Warning: Training directory for class '{cls}' not found at {class_path}. Assuming 0 samples.")
        train_samples_count[cls] = count


    print(f"Training quadrant counts per class: {train_samples_count}")
    # print(f"Class to label mapping used for weights: {class_label_map}") # Already printed class names

    if not train_files_found or not all_train_labels:
         print("Warning: No training quadrant samples found after generation. Cannot calculate class weights.")
         class_weights = None
    else:
        # Calculate weights using sklearn.utils.class_weight
        unique_labels = sorted(np.unique(all_train_labels)) # Get unique labels present in training data
        print(f"Unique label indices found in training data: {unique_labels}")
        # Map indices back to names for clarity
        present_classes = [class_names[i] for i in unique_labels if i < len(class_names)] # Added bounds check
        print(f"Classes present in training data: {present_classes}")


        if len(unique_labels) < num_classes:
            print(f"Warning: Only {len(unique_labels)} out of {num_classes} classes are present in the training set.")
            missing_classes = [cls for i, cls in enumerate(class_names) if i not in unique_labels]
            print(f"Missing classes: {missing_classes}")
            # Proceed to calculate weights only for present classes

        # 'balanced' mode automatically calculates weights inversely proportional to class frequencies
        class_weights_array = class_weight.compute_class_weight(
            class_weight='balanced', # Specify 'balanced' explicitly
            classes=unique_labels, # Use only the labels actually present
            y=all_train_labels     # Provide all the labels
        )

        # Create the class_weight dictionary expected by Keras fit method
        # Map the calculated weight to the correct class index (0 to num_classes-1)
        class_weights = {i: 0.0 for i in range(num_classes)} # Initialize all weights to 0
        for i, label_index in enumerate(unique_labels):
             if label_index < num_classes: # Safety check
                 class_weights[label_index] = class_weights_array[i]
             else:
                 print(f"Warning: Label index {label_index} out of bounds (num_classes={num_classes}). Skipping weight assignment.")

        print(f"Calculated Class Weights (index: weight): {class_weights}")

except Exception as e:
     print(f"An unexpected error occurred during class weight calculation: {e}")
     class_weights = None

if class_weights is None:
    print("Proceeding without class weights.")


# --- 5. Data Augmentation and Generators (for Quadrants) ---
print("\n--- Setting up Data Generators for Quadrants ---")

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest',
    brightness_range=[0.85, 1.15]
    )

# Validation and Test generators should only rescale
validation_test_datagen = ImageDataGenerator(rescale=1./255)

# Initialize generators to None for error checking
train_generator, validation_generator, test_generator = None, None, None

try:
    train_generator = train_datagen.flow_from_directory(
        train_dir,
        target_size=(QUAD_IMG_HEIGHT, QUAD_IMG_WIDTH),
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        classes=class_names, # Ensure consistent class order
        color_mode='rgb', # Explicitly set color mode
        shuffle=True,
        seed=RANDOM_STATE) # Add seed for reproducibility if needed

    validation_generator = validation_test_datagen.flow_from_directory(
        validation_dir,
        target_size=(QUAD_IMG_HEIGHT, QUAD_IMG_WIDTH),
        batch_size=BATCH_SIZE, # Can use larger batch size for validation if memory allows
        class_mode='categorical',
        classes=class_names,
        color_mode='rgb',
        shuffle=False) # No need to shuffle validation

    test_generator = validation_test_datagen.flow_from_directory(
        test_dir,
        target_size=(QUAD_IMG_HEIGHT, QUAD_IMG_WIDTH),
        batch_size=1, # Batch size 1 for precise evaluation
        class_mode='categorical',
        classes=class_names,
        color_mode='rgb',
        shuffle=False) # DO NOT shuffle test data

    print(f"Found {train_generator.samples} training quadrant images belonging to {train_generator.num_classes} classes.")
    print(f"Found {validation_generator.samples} validation quadrant images belonging to {validation_generator.num_classes} classes.")
    print(f"Found {test_generator.samples} test quadrant images belonging to {test_generator.num_classes} classes.")

    # Crucial Check: Ensure generators actually found images
    if train_generator.samples == 0:
         print("CRITICAL WARNING: Training generator found 0 images. Check 'train_dir' contents and permissions.")
         # Stop if no training data
         raise RuntimeError("No training images found by the generator.")
    if validation_generator.samples == 0:
         print("Warning: Validation generator found 0 images. Validation metrics will be unavailable. Early stopping might behave unexpectedly.")
    if test_generator.samples == 0:
          print("Warning: Test generator found 0 images. Final evaluation on test set will be skipped.")


    print(f"Class Indices mapping used by generators: {train_generator.class_indices}")

    # Verify class indices match across generators (important!)
    # Check only if all generators were created successfully
    if train_generator and validation_generator and test_generator:
        if not (train_generator.class_indices == validation_generator.class_indices == test_generator.class_indices):
            print("CRITICAL WARNING: Class indices mismatch between generators!")
            print(f"Train: {train_generator.class_indices}")
            print(f"Val: {validation_generator.class_indices}")
            print(f"Test: {test_generator.class_indices}")
            # This usually indicates inconsistent subdirectories or issues with the 'classes' parameter.
            # Stop execution if this happens.
            raise RuntimeError("Class indices mismatch in data generators.")
        else:
            # Store generator class names for later reports (using test_generator's mapping)
            generator_class_names = list(test_generator.class_indices.keys()) # Should match 'class_names'
    elif train_generator: # If only train generator exists
        generator_class_names = list(train_generator.class_indices.keys())
    else: # Should have exited if train_generator is None
        generator_class_names = class_names # Fallback


except FileNotFoundError as e:
     print(f"Error initializing ImageDataGenerator: {e}. Check that directories exist and are populated:")
     print(f"Train Dir: {train_dir} (Exists: {os.path.exists(train_dir)})")
     print(f"Validation Dir: {validation_dir} (Exists: {os.path.exists(validation_dir)})")
     print(f"Test Dir: {test_dir} (Exists: {os.path.exists(test_dir)})")
     exit(1)
except RuntimeError as e: # Catch the explicit runtime errors raised above
      print(f"Runtime error during generator setup: {e}")
      exit(1)
except Exception as e:
     print(f"An unexpected error occurred setting up data generators: {e}")
     exit(1)

# --- 6. Model Building (for Quadrants) ---
print("\n--- Building Model for Quadrants (Stage 1: Train Head) ---")
model = None # Initialize model to None

# Ensure correct input shape with 3 channels
if QUAD_IMG_SHAPE[-1] != 3:
    print(f"Warning: QUAD_IMG_SHAPE {QUAD_IMG_SHAPE} doesn't have 3 channels. Adjusting for MobileNetV2.")
    QUAD_IMG_SHAPE = (QUAD_IMG_HEIGHT, QUAD_IMG_WIDTH, 3)

try:
    # Instantiate the base model
    base_model = MobileNetV2(input_shape=QUAD_IMG_SHAPE,
                             include_top=False, # Exclude the final classification layer
                             weights='imagenet') # Load weights pre-trained on ImageNet

    # Stage 1: Freeze the base model layers
    base_model.trainable = False
    print(f"Base model '{base_model.name}' loaded. Trainable status (Stage 1): {base_model.trainable}")
    # print(f"Number of layers in base model: {len(base_model.layers)}")

    # Build the new classification head
    inputs = Input(shape=QUAD_IMG_SHAPE)
    # We need to make sure the input provided to the base_model is compatible.
    # MobileNetV2 expects inputs to be scaled in a specific way if using tf.keras.applications.mobilenet_v2.preprocess_input
    # However, since we rescale with 1./255 in the generator, we don't need the specific preprocess_input layer here.
    x = base_model(inputs, training=False) # IMPORTANT: training=False for stage 1 keeps BatchNorm stats frozen
    x = GlobalAveragePooling2D(name="global_avg_pool")(x) # Pool features spatially
    x = Dropout(DROPOUT_RATE, name="head_dropout")(x) # Regularization
    x = Dense(DENSE_UNITS, activation='relu', name="head_dense")(x) # Intermediate dense layer
    outputs = Dense(num_classes, activation='softmax', name="head_softmax")(x) # Output layer for classification

    model = Model(inputs, outputs)

    # Compile for Stage 1 (Training the head)
    model.compile(optimizer=Adam(learning_rate=INITIAL_LEARNING_RATE),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    print("\nModel Summary (Stage 1 - Head Training):")
    model.summary()

except Exception as e:
    print(f"Error building the model: {e}")
    exit(1)


# --- 7. Model Training (Stage 1) ---
print("\n--- Starting Model Training (Stage 1: Head Only) ---")

early_stopping_stage1 = EarlyStopping(
    monitor='val_loss',       # Monitor validation loss
    patience=EARLY_STOP_PATIENCE_STAGE1, # Wait N epochs for improvement
    restore_best_weights=True, # Restore weights from the epoch with the best val_loss
    verbose=1,                # Print messages when stopping
    mode='min'                # Stop when monitored quantity stops decreasing
)

# Calculate steps, ensuring they are at least 1 if data exists, handle 0 samples
steps_per_epoch = 0
if train_generator:
     steps_per_epoch = max(1, train_generator.samples // BATCH_SIZE) if train_generator.samples > 0 else 0

validation_steps = 0
monitor_metric = 'val_loss' # Default monitor metric
if validation_generator:
     validation_steps = max(1, validation_generator.samples // BATCH_SIZE) if validation_generator.samples > 0 else 0
     if validation_steps == 0:
          print("Warning: Validation generator has 0 samples. Early stopping will monitor 'loss' instead of 'val_loss'.")
          monitor_metric = 'loss'
          early_stopping_stage1 = EarlyStopping(monitor=monitor_metric, patience=EARLY_STOP_PATIENCE_STAGE1, restore_best_weights=True, verbose=1, mode='min')

history_stage1 = None
if model is None:
     print("CRITICAL ERROR: Model was not built successfully. Cannot start training.")
     exit(1)
elif steps_per_epoch == 0:
     print("CRITICAL ERROR: No training samples found by the generator. Cannot start Stage 1 training.")
     exit(1)
else:
    print(f"Stage 1 Training with: steps_per_epoch={steps_per_epoch}, validation_steps={validation_steps}")
    print(f"Monitoring metric for Early Stopping: {monitor_metric}")
    print(f"Using class weights: {class_weights is not None}")
    try:
        fit_kwargs = {
            'steps_per_epoch': steps_per_epoch,
            'epochs': EPOCHS,
            'callbacks': [early_stopping_stage1],
            'class_weight': class_weights if class_weights is not None else None
        }
        # Only pass validation data if it exists and has steps
        if validation_generator and validation_steps > 0:
            fit_kwargs['validation_data'] = validation_generator
            fit_kwargs['validation_steps'] = validation_steps

        history_stage1 = model.fit(
            train_generator,
            **fit_kwargs
        )
        print("\n--- Model Training Stage 1 Complete ---")
        # Optional: Save weights after stage 1
        # model.save_weights('/tmp/quadrant_model_stage1_weights.h5')
        # print("Saved model weights after Stage 1.")
    except Exception as e:
        print(f"Error during Stage 1 model training: {e}")
        # Decide if you want to proceed without stage 1 results
        # exit(1) # Option to exit if stage 1 fails

# --- 8. Model Fine-Tuning (Stage 2) ---
history_stage2 = None # Initialize
fine_tuning_performed = False

can_fine_tune = True
if history_stage1 is None:
    print("\n--- Skipping Fine-Tuning: Stage 1 did not run or failed ---")
    can_fine_tune = False
elif model is None: # Should not happen if Stage 1 ran, but check anyway
    print("\n--- Skipping Fine-Tuning: Model object is not available ---")
    can_fine_tune = False
elif FINE_TUNE_AT_LAYER <= 0:
    print("\n--- Skipping Fine-Tuning: FINE_TUNE_AT_LAYER set to 0 or less ---")
    can_fine_tune = False
elif not validation_generator or validation_steps == 0: # Cannot effectively fine-tune without validation data
     print("\n--- Skipping Fine-Tuning: Validation data was missing or empty, cannot monitor fine-tuning effectively. ---")
     can_fine_tune = False

if can_fine_tune:
    print("\n--- Preparing for Fine-Tuning (Stage 2) ---")

    # Unfreeze layers in the base model
    base_model.trainable = True
    print(f"Base model trainable status set to: {base_model.trainable}")
    # print(f"Total layers in base model: {len(base_model.layers)}")

    # Freeze layers *before* FINE_TUNE_AT_LAYER
    num_layers_total = len(base_model.layers)
    if 0 < FINE_TUNE_AT_LAYER < num_layers_total:
        print(f"Freezing the first {FINE_TUNE_AT_LAYER} layers of the base model.")
        for layer in base_model.layers[:FINE_TUNE_AT_LAYER]:
            layer.trainable = False
        # Verify
        num_trainable_weights = len(model.trainable_weights)
        num_frozen_base_layers = sum(1 for layer in base_model.layers if not layer.trainable)
        print(f"Number of frozen layers in base model: {num_frozen_base_layers}")
        print(f"Total trainable weights in model after partial unfreeze: {num_trainable_weights}")

    elif FINE_TUNE_AT_LAYER >= num_layers_total:
        print(f"Warning: FINE_TUNE_AT_LAYER ({FINE_TUNE_AT_LAYER}) is >= number of layers ({num_layers_total}). Unfreezing ALL base model layers.")
        # All layers remain trainable (as set by base_model.trainable = True)
        num_trainable_weights = len(model.trainable_weights)
        print(f"Total trainable weights in model (all base layers unfrozen): {num_trainable_weights}")
    else: # FINE_TUNE_AT_LAYER is likely 0 or negative (already handled above, but for completeness)
         print("Warning: Invalid FINE_TUNE_AT_LAYER. Assuming full base model unfreezing.")


    # Re-compile the model with a *very low* learning rate for fine-tuning
    # Use a new optimizer instance
    fine_tune_optimizer = Adam(learning_rate=FINE_TUNE_LEARNING_RATE)
    model.compile(optimizer=fine_tune_optimizer, # Crucial: Use a very low LR
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    print("\nModel Summary (Stage 2 - Fine-Tuning):")
    model.summary() # Note the change in trainable parameters

    print("\n--- Starting Model Training (Stage 2: Fine-Tuning) ---")
    early_stopping_stage2 = EarlyStopping(
        monitor='val_loss', # Monitor val_loss again, as we have validation data now
        patience=EARLY_STOP_PATIENCE_STAGE2, # Potentially different patience
        restore_best_weights=True, # Restore best weights found during fine-tuning
        verbose=1,
        mode='min'
    )

    # Determine the initial epoch for fine-tuning history to make plots continuous (optional)
    initial_epoch_ft = 0
    epochs_stage1 = 0
    if history_stage1 and hasattr(history_stage1, 'epoch'):
         epochs_stage1 = len(history_stage1.epoch)
         # If stage 1 stopped early via callback, its history might reflect that already
         # Check if restore_best_weights was True for stage 1
         if early_stopping_stage1.restore_best_weights:
              # The loaded weights are from the best epoch, but history continues.
              # For plotting continuity, we just number epochs sequentially.
              initial_epoch_ft = epochs_stage1
              print(f"Stage 1 ran for {epochs_stage1} epochs (best weights restored). Starting fine-tuning epoch numbering from {initial_epoch_ft}.")
         else:
              # If best weights weren't restored, model state is from the last epoch run.
              initial_epoch_ft = epochs_stage1
              print(f"Stage 1 ran for {epochs_stage1} epochs. Starting fine-tuning epoch numbering from {initial_epoch_ft}.")

    # Fine-tune for potentially fewer epochs, or rely heavily on early stopping
    # Let's use the same total EPOCHS limit, but EarlyStopping is key.
    total_epochs_combined = EPOCHS # Or set a new limit like initial_epoch_ft + (EPOCHS // 2) for fine-tuning phase

    try:
        # Use the same fit_kwargs logic as stage 1, ensuring validation data is passed
        fit_kwargs_ft = {
            'steps_per_epoch': steps_per_epoch,
            'epochs': total_epochs_combined, # Max epochs across both stages
            'initial_epoch': initial_epoch_ft, # Start epoch numbering correctly
            'callbacks': [early_stopping_stage2],
            'class_weight': class_weights if class_weights is not None else None,
             # We know validation_generator and validation_steps exist and are > 0 here
            'validation_data': validation_generator,
            'validation_steps': validation_steps,
        }

        history_stage2 = model.fit(
            train_generator,
            **fit_kwargs_ft
        )
        print("\n--- Model Training Stage 2 (Fine-Tuning) Complete ---")
        fine_tuning_performed = True # Mark that this stage ran
    except Exception as e:
         print(f"Error during Stage 2 (Fine-Tuning) model training: {e}")
         # Model will retain weights restored from Stage 1's EarlyStopping if stage 2 failed


# --- 9. Evaluation (on Test Quadrants) ---
# Evaluate the final model (best weights restored by EarlyStopping from the last successfully run stage)

print("\n--- Evaluating Final Quadrant Model ---")

# Determine which history to plot and report on
final_history = None
training_stage_name = "N/A"
best_epoch = -1
best_epoch_idx = -1 # Store index for accessing metric values

# Check if model exists before proceeding
if model is None:
     print("\n--- Skipping Evaluation: Model object does not exist. ---")
else:
    # Logic to determine final state and history (same as before)
    if fine_tuning_performed and history_stage2 is not None and hasattr(history_stage2, 'history') and history_stage2.history:
        final_history = history_stage2
        training_stage_name = "Fine-Tuning (Stage 2)"
        print("Evaluating model after Fine-Tuning stage.")
        if 'val_loss' in final_history.history and len(final_history.history['val_loss']) > 0:
            best_epoch_idx = np.argmin(final_history.history['val_loss'])
            best_val_loss = final_history.history['val_loss'][best_epoch_idx]
            best_epoch = final_history.epoch[best_epoch_idx]
            print(f"Best validation loss during fine-tuning: {best_val_loss:.4f} at epoch {best_epoch}")
    elif history_stage1 is not None and hasattr(history_stage1, 'history') and history_stage1.history:
        final_history = history_stage1
        training_stage_name = "Head Training (Stage 1)"
        print("Evaluating model after Head Training stage (Fine-tuning skipped or failed).")
        if 'val_loss' in final_history.history and len(final_history.history['val_loss']) > 0:
            best_epoch_idx = np.argmin(final_history.history['val_loss'])
            best_val_loss = final_history.history['val_loss'][best_epoch_idx]
            best_epoch = final_history.epoch[best_epoch_idx]
            print(f"Best validation loss during head training: {best_val_loss:.4f} at epoch {best_epoch}")
        elif 'loss' in final_history.history and len(final_history.history['loss']) > 0: # Fallback if only training loss available
            best_epoch_idx = np.argmin(final_history.history['loss'])
            best_loss = final_history.history['loss'][best_epoch_idx]
            best_epoch = final_history.epoch[best_epoch_idx]
            print(f"Best training loss during head training: {best_loss:.4f} at epoch {best_epoch} (monitored train loss)")
    else:
        print("\n--- Skipping Evaluation: Model was not trained successfully in any stage or history is empty. ---")
        final_history = None # Ensure it's None

    if final_history is not None:
        # Plot Training history of the final stage that ran
        acc = final_history.history.get('accuracy', [])
        val_acc = final_history.history.get('val_accuracy', [])
        loss = final_history.history.get('loss', [])
        val_loss = final_history.history.get('val_loss', [])

        # Use the epoch numbers recorded in the history object
        epochs_run = final_history.epoch

        plt.figure(figsize=(12, 5))

        plot_created = False
        # Plot Accuracy if available
        if acc and val_acc:
            plt.subplot(1, 2, 1)
            plt.plot(epochs_run, acc, label='Training Accuracy')
            plt.plot(epochs_run, val_acc, label='Validation Accuracy')
            if best_epoch != -1 and best_epoch_idx != -1 and val_acc: # Mark best val loss epoch on accuracy plot
                plt.scatter(best_epoch, val_acc[best_epoch_idx],
                            color='red', label=f'Best Val Loss Epoch ({best_epoch})', zorder=5)
            plt.legend(loc='best')
            plt.title(f'Quadrant Accuracy ({training_stage_name})')
            plt.xlabel('Epoch')
            plt.ylabel('Accuracy')
            plot_created = True
        elif acc: # Only training accuracy available
             plt.subplot(1, 2, 1)
             plt.plot(epochs_run, acc, label='Training Accuracy')
             if best_epoch != -1 and best_epoch_idx != -1: # Mark best train loss epoch
                plt.scatter(best_epoch, acc[best_epoch_idx],
                         color='red', label=f'Best Train Loss Epoch ({best_epoch})', zorder=5)
             plt.legend(loc='best')
             plt.title(f'Quadrant Training Accuracy ({training_stage_name})')
             plt.xlabel('Epoch')
             plt.ylabel('Accuracy')
             plot_created = True


        # Plot Loss if available
        loss_plot_idx = 2 if plot_created else 1 # Adjust subplot index if accuracy wasn't plotted
        if loss and val_loss:
            plt.subplot(1, loss_plot_idx + 1, loss_plot_idx) # Use 1 row, 1 or 2 cols
            plt.plot(epochs_run, loss, label='Training Loss')
            plt.plot(epochs_run, val_loss, label='Validation Loss')
            if best_epoch != -1 and best_epoch_idx != -1: # Mark best val loss epoch
                plt.scatter(best_epoch, val_loss[best_epoch_idx],
                            color='red', label=f'Best Val Loss Epoch ({best_epoch})', zorder=5)
            plt.legend(loc='best')
            plt.title(f'Quadrant Model Loss ({training_stage_name})')
            plt.xlabel('Epoch')
            plt.ylabel('Loss')
            plot_created = True
        elif loss: # Only training loss available
             plt.subplot(1, loss_plot_idx + 1, loss_plot_idx)
             plt.plot(epochs_run, loss, label='Training Loss')
             if best_epoch != -1 and best_epoch_idx != -1: # Mark best train loss epoch
                plt.scatter(best_epoch, loss[best_epoch_idx],
                         color='red', label=f'Best Train Loss Epoch ({best_epoch})', zorder=5)
             plt.legend(loc='best')
             plt.title(f'Quadrant Training Loss ({training_stage_name})')
             plt.xlabel('Epoch')
             plt.ylabel('Loss')
             plot_created = True


        if plot_created:
            plt.tight_layout()
            plot_filename = "quadrant_training_history_final_stage.png"
            plt.savefig(plot_filename)
            print(f"Saved final stage training history plot to {plot_filename}")
            plt.show()
        else:
             print("No training history metrics found to plot.")

        # Evaluate on Test Set (Quadrants) using the final model state (best weights restored)
        print("\nEvaluating final model on Test Set Quadrants...")
        if test_generator is None or test_generator.samples == 0:
            print("Warning: No test quadrant samples found by generator. Cannot evaluate on test set.")
        else:
            test_steps = test_generator.samples # Since batch_size is 1
            try:
                # Ensure generator is reset before evaluation
                test_generator.reset()
                test_loss, test_acc = model.evaluate(test_generator, steps=test_steps, verbose=1)
                print(f"Test Quadrant Loss: {test_loss:.4f}")
                print(f"Test Quadrant Accuracy: {test_acc:.4f}")

                # Predictions and Detailed Metrics (on Quadrants)
                print("\nGenerating Predictions for Detailed Metrics on Test Quadrants...")
                test_generator.reset() # Reset again before prediction
                # Ensure steps match the number of samples when batch_size is 1
                Y_pred_probs = model.predict(test_generator, steps=test_generator.samples, verbose=1)
                y_pred_indices = np.argmax(Y_pred_probs, axis=1)

                # Get true labels, ensuring the length matches predictions
                y_true_indices = test_generator.classes
                if len(y_pred_indices) != len(y_true_indices):
                    print(f"CRITICAL WARNING: Mismatch between number of predictions ({len(y_pred_indices)}) and true labels ({len(y_true_indices)}). Evaluation metrics will be incorrect.")
                    # Attempt to truncate - but this indicates a deeper problem.
                    min_len = min(len(y_pred_indices), len(y_true_indices))
                    print(f"Truncating both prediction and true labels to length {min_len}.")
                    y_pred_indices = y_pred_indices[:min_len]
                    y_true_indices = y_true_indices[:min_len]


                print(f"\nClasses used for report/matrix (from generator): {generator_class_names}")

                print("\nClassification Report (on Test Quadrants):")
                # Use labels=range(num_classes) to ensure all potential classes appear in the report
                # Use generator_class_names which should align with the labels range
                print(classification_report(y_true_indices, y_pred_indices, labels=range(num_classes), target_names=generator_class_names, zero_division=0))

                print("\nConfusion Matrix (on Test Quadrants):")
                cm = confusion_matrix(y_true_indices, y_pred_indices, labels=range(num_classes)) # Ensure matrix includes all classes
                plt.figure(figsize=(10, 8))
                sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                            xticklabels=generator_class_names, yticklabels=generator_class_names)
                plt.title('Confusion Matrix (Test Quadrants)')
                plt.ylabel('True Label')
                plt.xlabel('Predicted Label')
                cm_filename = "quadrant_confusion_matrix.png"
                plt.savefig(cm_filename)
                print(f"Saved confusion matrix plot to {cm_filename}")
                plt.show()

                # ROC Curves (Optional, per-quadrant interpretation)
                try:
                    y_true_bin = label_binarize(y_true_indices, classes=range(num_classes))
                    n_classes_roc = y_true_bin.shape[1]

                    if n_classes_roc > 1 and Y_pred_probs.shape[1] == n_classes_roc: # Check if multiclass and predictions match
                        fpr = dict()
                        tpr = dict()
                        roc_auc = dict()
                        plt.figure(figsize=(8, 6))
                        colors = cycle(['aqua', 'darkorange', 'cornflowerblue', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan'])
                        roc_plot_created = False

                        for i, color in zip(range(n_classes_roc), colors):
                            # Ensure y_true_bin has samples for this class before calculating ROC
                            # Also check if class is present in predictions (though less likely to be missing if present in true)
                            if np.sum(y_true_bin[:, i]) > 0:
                                fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], Y_pred_probs[:, i])
                                roc_auc[i] = auc(fpr[i], tpr[i])
                                plt.plot(fpr[i], tpr[i], color=color, lw=2,
                                        label='ROC curve of class {0} (area = {1:0.2f})'
                                        ''.format(generator_class_names[i], roc_auc[i] if roc_auc[i] is not None else 0.0))
                                roc_plot_created = True
                            else:
                                fpr[i], tpr[i], roc_auc[i] = None, None, None # Class not present in y_true
                                # print(f"Skipping ROC plot for class '{generator_class_names[i]}' (no true samples in test set).") # Less verbose

                        if roc_plot_created:
                            plt.plot([0, 1], [0, 1], 'k--', lw=2)
                            plt.xlim([0.0, 1.0])
                            plt.ylim([0.0, 1.05])
                            plt.xlabel('False Positive Rate')
                            plt.ylabel('True Positive Rate')
                            plt.title('Multi-class ROC (Test Quadrants)')
                            plt.legend(loc="lower right")
                            roc_filename = "quadrant_roc_curves.png"
                            plt.savefig(roc_filename)
                            print(f"Saved ROC curve plot to {roc_filename}")
                            plt.show()
                        else:
                             print("No ROC curves plotted (possibly due to no positive samples for any class in test set).")

                    elif n_classes_roc <= 1:
                         print("Skipping ROC curve generation (only one class or binary case not fully handled here).")
                    else: # Prediction shape mismatch
                         print(f"Skipping ROC curve generation (prediction probability shape {Y_pred_probs.shape} mismatch with number of classes {n_classes_roc}).")

                except ValueError as roc_val_err:
                     print(f"Could not generate ROC curves due to ValueError: {roc_val_err}")
                except Exception as roc_e:
                     print(f"Could not generate ROC curves due to unexpected error: {roc_e}")


            except Exception as e:
                print(f"Error during test set evaluation or metrics generation: {e}")


# --- 10. Inference Strategy for a Full OPG (Example) ---
def predict_full_opg(image_path, model, target_quad_size=(QUAD_IMG_WIDTH, QUAD_IMG_HEIGHT), class_names_list=class_names):
    """
    Predicts the class for a full OPG by processing its quadrants.
    Uses averaging of probabilities as the combination strategy.
    Returns: predicted class name, averaged probabilities array, dictionary of quadrant probabilities
    """
    quadrant_predictions = {}
    if not os.path.exists(image_path):
        print(f"Error: Image path does not exist: {image_path}")
        return None, None, None
    if model is None:
         print("Error: Model object is None in predict_full_opg.")
         return None, None, None

    try:
        img = cv2.imread(image_path)
        if img is None: raise IOError(f"Cannot read image file (may be corrupted or format not supported by OpenCV): {image_path}")

        # Ensure 3 channels (BGR)
        if len(img.shape) == 2 or img.shape[2] == 1: img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
        elif img.shape[2] == 4: img = cv2.cvtColor(img, cv2.COLOR_BGRA2BGR)
        elif img.shape[2] != 3: raise ValueError(f"Unsupported number of channels ({img.shape[2]}) in image {image_path}")


        h, w, _ = img.shape
        if h < 2 or w < 2: raise ValueError(f"Image too small ({w}x{h}) to split into quadrants: {image_path}")
        h_mid, w_mid = h // 2, w // 2

        quadrants_data = []
        quad_names = ["q1_tl", "q2_tr", "q3_bl", "q4_br"]
        quadrant_coords = [
            (0, h_mid, 0, w_mid), (0, h_mid, w_mid, w),
            (h_mid, h, 0, w_mid), (h_mid, h, w_mid, w)
        ]

        valid_quad_imgs = []
        valid_quad_names = []

        for name, (r1, r2, c1, c2) in zip(quad_names, quadrant_coords):
            q_img = img[r1:r2, c1:c2]
            if q_img.size > 0:
                 # Resize quadrant
                 if q_img.shape[0] > target_quad_size[1] or q_img.shape[1] > target_quad_size[0]:
                     interpolation = cv2.INTER_AREA
                 else:
                     interpolation = cv2.INTER_LINEAR
                 q_img_resized = cv2.resize(q_img, target_quad_size, interpolation=interpolation)

                 # Rescale like the generator (important!)
                 q_img_rescaled = q_img_resized / 255.0
                 valid_quad_imgs.append(q_img_rescaled)
                 valid_quad_names.append(name)
            # else:
            #     print(f"Debug: Skipping empty quadrant {name} for {image_path}")


        if not valid_quad_imgs:
            print(f"Error: No valid (non-empty) quadrants extracted from {image_path}")
            return None, None, None

        # Predict on all valid quadrants in a batch
        quad_batch = np.array(valid_quad_imgs)
        # Verify batch shape matches model input expectations
        expected_input_shape = model.input_shape[1:] # Get shape without batch dimension
        if quad_batch.shape[1:] != expected_input_shape:
             raise ValueError(f"Quadrant batch shape mismatch. Expected {(None,) + expected_input_shape}, got {quad_batch.shape}")

        predictions = model.predict(quad_batch) # Get probabilities for each quadrant (shape: num_valid_quads, num_classes)

        # Store individual quadrant predictions
        for name, probs in zip(valid_quad_names, predictions):
            quadrant_predictions[name] = {cn: p for cn, p in zip(class_names_list, probs)}


        # Combine predictions - Average Probability Strategy
        if predictions.shape[0] > 0:
            avg_probabilities = np.mean(predictions, axis=0) # Average across the valid quadrants (axis 0)
            final_pred_index = np.argmax(avg_probabilities)
            final_pred_class = class_names_list[final_pred_index]
            # final_pred_confidence = avg_probabilities[final_pred_index] # Max of the averaged probs

            print(f"\n--- Prediction for Full OPG: {os.path.basename(image_path)} ---")
            print("Probabilities per Quadrant:")
            for q_name, q_probs_dict in quadrant_predictions.items():
                print(f"  {q_name}: { {k: f'{v:.3f}' for k,v in q_probs_dict.items()} }") # Formatted output
            print(f"Averaged probabilities across {len(valid_quad_names)} quadrants: { {cn: f'{p:.3f}' for cn, p in zip(class_names_list, avg_probabilities)} }")
            print(f"==> Final Combined Prediction: {final_pred_class} (Based on highest average probability)")
            print("--- End Prediction ---")

            return final_pred_class, avg_probabilities, quadrant_predictions
        else: # Should not happen if valid_quad_imgs is not empty, but as a safeguard
             print(f"Error: No predictions generated despite having valid quadrants for {image_path}")
             return None, None, quadrant_predictions


    except IOError as e:
        print(f"I/O Error during full OPG prediction for {image_path}: {e}")
        return None, None, None
    except ValueError as e:
         print(f"Value Error during full OPG prediction for {image_path}: {e}")
         return None, None, None
    except cv2.error as cv_err:
         print(f"OpenCV Error during full OPG prediction for {image_path}: {cv_err}")
         return None, None, None
    except Exception as e:
        print(f"An unexpected error occurred during full OPG prediction for {image_path}: {e.__class__.__name__}: {e}")
        return None, None, None

# Example Usage (after model training/evaluation):
print("\n--- Example: Predicting on a Full Original OPG from Test Set Split ---")

# Check if the model exists and if we have the list of original test files
if 'model' in locals() and model is not None and 'original_test_files' in locals() and original_test_files:
    # Try to find one original test image to predict
    example_img_path = None
    # Flatten the dictionary of test files lists {class: [path1, path2], ...} -> [path1, path2, ...]
    all_orig_test_paths = [item for sublist in original_test_files.values() for item in sublist if isinstance(sublist, list)] # Added check for list

    if all_orig_test_paths:
        # Select a random test file for variety, if possible
        try:
             example_img_path = np.random.choice(all_orig_test_paths)
        except: # Fallback if random choice fails
             example_img_path = all_orig_test_paths[0]

        print(f"Using an original test image identified during split for prediction: {example_img_path}")
        predicted_class, predicted_probs, quad_probs_dict = predict_full_opg(example_img_path, model, class_names_list=class_names)
        # You could add code here to compare predicted_class to the actual class derived from the example_img_path if needed
    else:
        print("Could not find any original test images recorded during the split process to use for the example.")
        # Fallback: Try grabbing *any* image from raw data (less representative of test performance)
        print("Attempting fallback: using an image from the raw data directory...")
        example_cls_fallback = class_names[0] if class_names else None # Take first class if available
        if example_cls_fallback:
            fallback_dir = os.path.join(raw_data_dir, example_cls_fallback)
            try:
                if os.path.isdir(fallback_dir):
                    example_files = glob.glob(os.path.join(fallback_dir, '*.*'))
                    if example_files:
                        example_img_path = example_files[0]
                        print(f"Using fallback example image: {example_img_path}")
                        predicted_class, predicted_probs, quad_probs_dict = predict_full_opg(example_img_path, model, class_names_list=class_names)
                    else:
                        print(f"Could not find any image files in fallback directory: {fallback_dir}")
                else:
                    print(f"Fallback directory does not exist: {fallback_dir}")
            except Exception as e:
                print(f"Error finding fallback example image: {e}")
        else:
             print("No classes defined, cannot select fallback directory.")


elif 'model' not in locals() or model is None:
     print("Model was not trained or loaded successfully. Skipping full OPG prediction example.")
else: # Model exists but original_test_files list is missing or empty
     print("Original test file list not available or empty. Skipping example prediction based on identified test set images.")


# --- Optional Cleanup ---
# Uncomment the following lines to remove the downloaded and processed data
# print(f"\n--- Optional Cleanup: Removing {base_local_dir} ---")
# try:
#     if os.path.exists(base_local_dir):
#         shutil.rmtree(base_local_dir)
#         print(f"Successfully removed {base_local_dir}")
#     else:
#         print(f"Directory {base_local_dir} not found, skipping cleanup.")
# except Exception as e:
#     print(f"Error during cleanup: {e}")

print("\n--- Script Execution Finished ---")
```

```python
# ******************************************************************************
# ** CRITICAL SETUP STEP FOR SAGEMAKER (If needed) **
# ******************************************************************************
# If you previously encountered the "libGL.so.1" error and haven't fixed it:
# 1. Open a Terminal in SageMaker (File -> New -> Terminal).
# 2. Run: sudo yum install -y mesa-libGL
# 3. Restart the Kernel (Kernel -> Restart Kernel).
# ******************************************************************************

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, Input
from tensorflow.keras.applications import MobileNetV2 # Or another suitable base model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
import cv2 # Using OpenCV for image manipulation (still needed for reading/resizing in prediction)

import sagemaker
import boto3
from urllib.parse import urlparse
import os
import shutil
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.utils import class_weight
from itertools import cycle
import glob
import time

# --- Basic Library Checks ---
print(f"TensorFlow Version: {tf.__version__}")
print(f"SageMaker SDK Version: {sagemaker.__version__}")
try:
    print(f"OpenCV Version: {cv2.__version__}")
except NameError:
    print("ERROR: OpenCV (cv2) could not be imported. Please ensure it and its dependencies (like libGL) are installed.")
    exit(1)

# --- 1. Configuration (Revised for Full OPGs) ---
s3_paths = {
    "Broken Teeth": "s3://sagemaker-studio-010526272250-hsf94lgtf6i/Mendeley_Data/Dental OPG (Classification)/BDC-BDR/",
    "Caries": "s3://sagemaker-studio-010526272250-hsf94lgtf6i/Mendeley_Data/Dental OPG (Classification)/Caries/",
    "Fractured Teeth": "s3://sagemaker-studio-010526272250-hsf94lgtf6i/Mendeley_Data/Dental OPG (Classification)/Fractured Teeth/",
    "Healthy Teeth": "s3://sagemaker-studio-010526272250-hsf94lgtf6i/Mendeley_Data/Dental OPG (Classification)/Healthy Teeth/",
    "Impacted Teeth": "s3://sagemaker-studio-010526272250-hsf94lgtf6i/Mendeley_Data/Dental OPG (Classification)/Impacted teeth/",
    "Infected Teeth": "s3://sagemaker-studio-010526272250-hsf94lgtf6i/Mendeley_Data/Dental OPG (Classification)/Infection/"
}

# --- Make class names filesystem-friendly and sort ---
class_names = sorted([name.replace(" ", "_") for name in s3_paths.keys()])
s3_paths_clean_keys = {name.replace(" ", "_"): path for name, path in s3_paths.items()}
num_classes = len(class_names)

print(f"Classes found: {class_names}")
print(f"Number of classes: {num_classes}")

# --- Local directories (Revised for Full OPGs) ---
base_local_dir_full = '/tmp/dental_data_full_opg' # New base directory
raw_data_dir = os.path.join(base_local_dir_full, 'raw')
processed_data_dir_full = os.path.join(base_local_dir_full, 'processed') # This will hold split full images
train_dir_full = os.path.join(processed_data_dir_full, 'train')
validation_dir_full = os.path.join(processed_data_dir_full, 'validation')
test_dir_full = os.path.join(processed_data_dir_full, 'test')

# --- Model/Training Parameters (Revised for Full OPGs) ---
IMG_HEIGHT = 224 # Standard input size for many models like MobileNetV2
IMG_WIDTH = 224
# Ensure 3 channels for pre-trained models
IMG_SHAPE = (IMG_HEIGHT, IMG_WIDTH, 3)

BATCH_SIZE = 32 # Adjust based on available GPU memory with larger images
EPOCHS = 50 # Rely on EarlyStopping
INITIAL_LEARNING_RATE = 1e-4
FINE_TUNE_LEARNING_RATE = 1e-5 # Keep fine-tuning LR low
VAL_SPLIT = 0.15 # Percentage of *original* data for validation
TEST_SPLIT = 0.15 # Percentage of *original* data for testing
RANDOM_STATE = 42 # For reproducible splits
DROPOUT_RATE = 0.4 # Can adjust later if needed
DENSE_UNITS = 256 # Can adjust later if needed
FINE_TUNE_AT_LAYER = 100 # Unfreeze layers from this index onwards for fine-tuning (adjust based on model)
EARLY_STOP_PATIENCE_STAGE1 = 10
EARLY_STOP_PATIENCE_STAGE2 = 10

# --- 2. Data Download (Same as before) ---

def download_s3_folder(s3_uri, local_dir):
    """Downloads files from an S3 prefix to a local directory."""
    if not os.path.exists(local_dir):
        os.makedirs(local_dir, exist_ok=True)
        print(f"Created directory: {local_dir}")
    else:
        print(f"Directory {local_dir} already exists. Contents might be overwritten or merged.")
        # Optional: Clean directory before download if needed for pristine state
        # print(f"Cleaning existing directory: {local_dir}")
        # shutil.rmtree(local_dir)
        # os.makedirs(local_dir, exist_ok=True)

    print(f"Downloading data from {s3_uri} to {local_dir}...")
    start_time = time.time()
    try:
        sagemaker.s3.S3Downloader.download(s3_uri, local_dir)
        elapsed_time = time.time() - start_time
        # Check if files were actually downloaded (simple check)
        if not os.listdir(local_dir):
             print(f"Warning: Download from {s3_uri} completed, but directory {local_dir} is empty. Check S3 path and permissions.")
        else:
             print(f"Successfully downloaded data from {s3_uri} in {elapsed_time:.2f} seconds")
    except Exception as e:
        print(f"Error downloading {s3_uri}: {e}")
        # Decide if the script should stop if download fails
        raise e # Reraise to stop execution

print("\n--- Starting Data Download ---")
# Clean raw_data_dir before downloading to ensure freshness
if os.path.exists(raw_data_dir):
    print(f"Removing existing raw data directory: {raw_data_dir}")
    shutil.rmtree(raw_data_dir)
os.makedirs(raw_data_dir) # Recreate after cleaning

download_success = True
for class_name in class_names:
    s3_path = s3_paths_clean_keys.get(class_name)
    if s3_path:
        local_class_dir = os.path.join(raw_data_dir, class_name)
        try:
            download_s3_folder(s3_path, local_class_dir)
        except Exception as e:
            print(f"Download failed for class '{class_name}', stopping execution.")
            download_success = False
            break # Stop if one class fails
    else:
        print(f"Warning: Could not find S3 path for class '{class_name}'.")
        # Consider if this is critical

if not download_success:
    print("\n--- Data Download Failed. Exiting. ---")
    exit(1)
else:
    print("\n--- Data Download Phase Complete ---")


# --- 3. Data Splitting (Full OPGs - Replaces Quadrant Generation) ---

def split_original_data(source_dir, train_dir, validation_dir, test_dir, classes, val_split, test_split, random_state=RANDOM_STATE):
    """
    Splits original image files from source_dir into train, validation, and test sets,
    copying the files into the respective target directories.
    """
    print(f"\n--- Starting Data Splitting for Full OPGs ---")

    # Ensure target directories exist and are empty
    for dir_path in [processed_data_dir_full, train_dir, validation_dir, test_dir]: # Include processed base dir
        if os.path.exists(dir_path):
            print(f"Removing existing directory: {dir_path}")
            shutil.rmtree(dir_path)
    # Recreate necessary directories
    os.makedirs(processed_data_dir_full)
    os.makedirs(train_dir)
    os.makedirs(validation_dir)
    os.makedirs(test_dir)
    for cls in classes:
        os.makedirs(os.path.join(train_dir, cls), exist_ok=True)
        os.makedirs(os.path.join(validation_dir, cls), exist_ok=True)
        os.makedirs(os.path.join(test_dir, cls), exist_ok=True)

    total_files_processed = 0
    split_summary = {}

    all_original_files = [] # List of (filepath, classname) tuples
    print(f"Scanning source directory: {source_dir}")
    for cls in classes:
        class_source_dir = os.path.join(source_dir, cls)
        if not os.path.isdir(class_source_dir):
             print(f"Warning: Source directory for class '{cls}' not found at {class_source_dir}. Skipping.")
             continue

        # Find original image files
        supported_extensions = ['*.png', '*.jpg', '*.jpeg', '*.bmp', '*.tif', '*.tiff']
        class_files = []
        for ext in supported_extensions:
            class_files.extend(glob.glob(os.path.join(class_source_dir, ext)))
            class_files.extend(glob.glob(os.path.join(class_source_dir, ext.upper())))

        # Add found files with their class label
        for f_path in sorted(list(set(class_files))): # Sort for consistency
            all_original_files.append((f_path, cls))

    if not all_original_files:
        print("CRITICAL ERROR: No image files found in the source directory after download.")
        raise ValueError("No source images found for splitting.")

    print(f"Found {len(all_original_files)} total original image files across all classes.")

    # Separate file paths and labels for splitting
    filepaths = [item[0] for item in all_original_files]
    labels = [item[1] for item in all_original_files]

    # Stratified split for train/val/test
    try:
        # Split off test set first
        if test_split > 0 and len(filepaths) > 1:
             train_val_paths, test_paths, train_val_labels, test_labels = train_test_split(
                filepaths, labels, test_size=test_split, random_state=random_state, stratify=labels
            )
        else:
             train_val_paths, test_paths = filepaths, []
             train_val_labels, test_labels = labels, []

        # Split off validation set from train_val
        if val_split > 0 and len(train_val_paths) > 1:
            # Adjust val_split relative to the remaining data
            relative_val_split = val_split / (1.0 - test_split) if (1.0 - test_split) > 0 else 0
            if relative_val_split >= 1.0 or len(train_val_paths) < 2: # Handle edge cases
                 train_paths, val_paths = train_val_paths, []
                 train_labels, val_labels = train_val_labels, []
            else:
                 train_paths, val_paths, train_labels, val_labels = train_test_split(
                    train_val_paths, train_val_labels, test_size=relative_val_split, random_state=random_state, stratify=train_val_labels
                )
        else:
            train_paths, val_paths = train_val_paths, []
            train_labels, val_labels = train_val_labels, []

    except ValueError as e:
        print(f"Warning: Stratified split failed ({e}). This might happen with very few samples per class. Falling back to non-stratified split.")
        # Non-stratified fallback (less ideal)
        train_val_paths, test_paths = train_test_split(filepaths, test_size=test_split, random_state=random_state)
        if val_split > 0 and len(train_val_paths) > 0:
             relative_val_split = val_split / (1.0 - test_split) if (1.0 - test_split) > 0 else 0
             if relative_val_split < 1.0:
                  train_paths, val_paths = train_test_split(train_val_paths, test_size=relative_val_split, random_state=random_state)
             else:
                   train_paths = train_val_paths
                   val_paths = []
        else:
             train_paths = train_val_paths
             val_paths = []
         # Note: Labels are not tracked in this non-stratified fallback easily, split_summary will just show counts.


    print(f"Splitting complete: {len(train_paths)} train, {len(val_paths)} validation, {len(test_paths)} test images.")

    # --- Copy files to destination directories ---
    def copy_files(file_list, dest_base_dir):
        counts = {cls: 0 for cls in classes}
        for src_path in file_list:
            try:
                filename = os.path.basename(src_path)
                # Find the class name directory from the source path (assumes structure raw/ClassName/file.jpg)
                class_name = os.path.basename(os.path.dirname(src_path))
                if class_name in classes:
                    dest_dir = os.path.join(dest_base_dir, class_name)
                    dest_path = os.path.join(dest_dir, filename)
                    # Ensure dest dir exists (it should, but double-check)
                    # os.makedirs(dest_dir, exist_ok=True) # Handled earlier
                    shutil.copy2(src_path, dest_path) # copy2 preserves metadata
                    counts[class_name] += 1
                else:
                     print(f"Warning: Could not determine class for {src_path}. Skipping copy.")
            except Exception as copy_e:
                print(f"Error copying {src_path} to {dest_base_dir}: {copy_e}")
        return counts

    print("Copying training files...")
    train_counts = copy_files(train_paths, train_dir)
    print("Copying validation files...")
    val_counts = copy_files(val_paths, validation_dir)
    print("Copying test files...")
    test_counts = copy_files(test_paths, test_dir)

    print("\n--- File Splitting Summary (Full OPGs) ---")
    total_copied = 0
    for cls in classes:
        tr_c = train_counts.get(cls, 0)
        v_c = val_counts.get(cls, 0)
        te_c = test_counts.get(cls, 0)
        print(f"Class '{cls}': Train={tr_c}, Val={v_c}, Test={te_c}")
        total_copied += tr_c + v_c + te_c
    print(f"Total Files Copied: {total_copied} (Should match total found if no errors)")
    print(f"--- File Splitting Complete ---")

    # Return the list of test file paths (useful for later evaluation)
    return test_paths


# --- Run the Splitting (for Full OPGs) ---
try:
    # Use the new directories for full OPGs
    original_test_filepaths = split_original_data(raw_data_dir, train_dir_full, validation_dir_full, test_dir_full, class_names, VAL_SPLIT, TEST_SPLIT)
except ValueError as e:
    print(f"Stopping script due to error during data preparation: {e}")
    exit(1)
except Exception as e:
     print(f"An unexpected error occurred during data preparation: {e}")
     exit(1)


# --- 4. Calculate Class Weights (Based on *Full OPG* counts in Train) ---
print("\n--- Calculating Class Weights for Training Set (Full OPGs) ---")
class_weights = None # Initialize
try:
    # Count samples per class in the *new* training directory
    train_samples_count = {}
    all_train_labels_indices = []
    class_label_map = {cls: i for i, cls in enumerate(class_names)} # Consistent mapping
    train_files_found = False

    print(f"Scanning training directory: {train_dir_full}")
    for cls, label_index in class_label_map.items():
        class_path = os.path.join(train_dir_full, cls)
        count = 0
        if os.path.isdir(class_path):
            try:
                files_in_class = [f for f in os.listdir(class_path) if os.path.isfile(os.path.join(class_path, f))]
                count = len(files_in_class)
            except OSError as list_err:
                print(f"Warning: Could not list files in {class_path}: {list_err}. Assuming 0 samples for {cls}.")
                count = 0

            if count > 0:
                train_files_found = True
                all_train_labels_indices.extend([label_index] * count)
        else:
             print(f"Warning: Training directory for class '{cls}' not found at {class_path}. Assuming 0 samples.")
        train_samples_count[cls] = count

    print(f"Training image counts per class: {train_samples_count}")

    if not train_files_found or not all_train_labels_indices:
         print("Warning: No training samples found after splitting. Cannot calculate class weights.")
         class_weights = None
    else:
        unique_labels = sorted(np.unique(all_train_labels_indices))
        print(f"Unique label indices found in training data: {unique_labels}")
        present_classes = [class_names[i] for i in unique_labels if i < len(class_names)]
        print(f"Classes present in training data: {present_classes}")

        if len(unique_labels) < num_classes:
            print(f"Warning: Only {len(unique_labels)} out of {num_classes} classes are present in the training set.")
            missing_classes = [cls for i, cls in enumerate(class_names) if i not in unique_labels]
            print(f"Missing classes: {missing_classes}")

        class_weights_array = class_weight.compute_class_weight(
            class_weight='balanced',
            classes=unique_labels,
            y=all_train_labels_indices
        )
        class_weights = {i: 0.0 for i in range(num_classes)}
        for i, label_index in enumerate(unique_labels):
             if label_index < num_classes:
                 class_weights[label_index] = class_weights_array[i]
             else:
                 print(f"Warning: Label index {label_index} out of bounds. Skipping weight.")
        print(f"Calculated Class Weights (index: weight): {class_weights}")

except Exception as e:
     print(f"An unexpected error occurred during class weight calculation: {e}")
     class_weights = None

if class_weights is None:
    print("Proceeding without class weights.")


# --- 5. Data Augmentation and Generators (for Full OPGs) ---
print("\n--- Setting up Data Generators for Full OPGs ---")

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=15,       # Adjust augmentation ranges if needed for larger images
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest',
    brightness_range=[0.9, 1.1] # Slightly tighter brightness range maybe
    )

validation_test_datagen = ImageDataGenerator(rescale=1./255) # Only rescale for val/test

# Initialize generators to None
train_generator, validation_generator, test_generator = None, None, None
generator_class_names = class_names # Default

try:
    train_generator = train_datagen.flow_from_directory(
        train_dir_full, # Use new train directory
        target_size=(IMG_HEIGHT, IMG_WIDTH), # Use full image dimensions
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        classes=class_names,
        color_mode='rgb',
        shuffle=True,
        seed=RANDOM_STATE)

    validation_generator = validation_test_datagen.flow_from_directory(
        validation_dir_full, # Use new validation directory
        target_size=(IMG_HEIGHT, IMG_WIDTH),
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        classes=class_names,
        color_mode='rgb',
        shuffle=False)

    test_generator = validation_test_datagen.flow_from_directory(
        test_dir_full, # Use new test directory
        target_size=(IMG_HEIGHT, IMG_WIDTH),
        batch_size=1, # Batch size 1 for evaluation
        class_mode='categorical',
        classes=class_names,
        color_mode='rgb',
        shuffle=False)

    print(f"Found {train_generator.samples} training images belonging to {train_generator.num_classes} classes.")
    print(f"Found {validation_generator.samples} validation images belonging to {validation_generator.num_classes} classes.")
    print(f"Found {test_generator.samples} test images belonging to {test_generator.num_classes} classes.")

    if train_generator.samples == 0:
         raise RuntimeError("No training images found by the generator.")
    if validation_generator.samples == 0:
         print("Warning: Validation generator found 0 images.")
    if test_generator.samples == 0:
          print("Warning: Test generator found 0 images.")

    print(f"Class Indices mapping: {train_generator.class_indices}")
    generator_class_names = list(train_generator.class_indices.keys()) # Store for reports

    # Verify class indices match (should if using same 'classes' list)
    if validation_generator and test_generator:
         if not (train_generator.class_indices == validation_generator.class_indices == test_generator.class_indices):
              raise RuntimeError("Class indices mismatch between generators!")

except FileNotFoundError as e:
     print(f"Error initializing ImageDataGenerator: {e}. Check directories:")
     print(f"Train: {train_dir_full} (Exists: {os.path.exists(train_dir_full)})")
     print(f"Val: {validation_dir_full} (Exists: {os.path.exists(validation_dir_full)})")
     print(f"Test: {test_dir_full} (Exists: {os.path.exists(test_dir_full)})")
     exit(1)
except RuntimeError as e:
      print(f"Runtime error during generator setup: {e}")
      exit(1)
except Exception as e:
     print(f"An unexpected error occurred setting up data generators: {e}")
     exit(1)


# --- 6. Model Building (for Full OPGs) ---
print("\n--- Building Model for Full OPGs ---")
model = None

try:
    # Base model with updated input shape
    base_model = MobileNetV2(input_shape=IMG_SHAPE, # Use full OPG shape
                             include_top=False,
                             weights='imagenet')

    # Stage 1: Freeze base model
    base_model.trainable = False
    print(f"Base model '{base_model.name}' loaded. Trainable status (Stage 1): {base_model.trainable}")

    # Build classification head (can keep the same structure initially)
    inputs = Input(shape=IMG_SHAPE)
    x = base_model(inputs, training=False) # training=False for stage 1
    x = GlobalAveragePooling2D(name="global_avg_pool")(x)
    x = Dropout(DROPOUT_RATE, name="head_dropout")(x)
    x = Dense(DENSE_UNITS, activation='relu', name="head_dense")(x)
    outputs = Dense(num_classes, activation='softmax', name="head_softmax")(x)

    model = Model(inputs, outputs)

    # Compile for Stage 1
    model.compile(optimizer=Adam(learning_rate=INITIAL_LEARNING_RATE),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    print("\nModel Summary (Stage 1 - Head Training):")
    model.summary()

except Exception as e:
    print(f"Error building the model: {e}")
    exit(1)


# --- 7. Model Training (Stage 1 - Full OPGs) ---
print("\n--- Starting Model Training (Stage 1: Head Only - Full OPGs) ---")

# Adjust monitor metric if validation data is missing
monitor_metric_s1 = 'val_loss'
validation_steps_s1 = 0
if validation_generator:
    validation_steps_s1 = max(1, validation_generator.samples // BATCH_SIZE) if validation_generator.samples > 0 else 0
    if validation_steps_s1 == 0:
        monitor_metric_s1 = 'loss'
        print("Warning: No validation samples for Stage 1. Monitoring 'loss' for EarlyStopping.")

early_stopping_stage1 = EarlyStopping(
    monitor=monitor_metric_s1,
    patience=EARLY_STOP_PATIENCE_STAGE1,
    restore_best_weights=True,
    verbose=1,
    mode='min'
)

steps_per_epoch_s1 = 0
if train_generator:
     steps_per_epoch_s1 = max(1, train_generator.samples // BATCH_SIZE) if train_generator.samples > 0 else 0

history_stage1 = None
if not model or steps_per_epoch_s1 == 0:
    print("Error: Model not built or no training data. Cannot start Stage 1.")
    exit(1)
else:
    print(f"Stage 1 Training with: steps_per_epoch={steps_per_epoch_s1}, validation_steps={validation_steps_s1}")
    print(f"Monitoring metric: {monitor_metric_s1}")
    print(f"Using class weights: {class_weights is not None}")
    try:
        fit_kwargs_s1 = {
            'steps_per_epoch': steps_per_epoch_s1,
            'epochs': EPOCHS,
            'callbacks': [early_stopping_stage1],
            'class_weight': class_weights if class_weights is not None else None
        }
        if validation_generator and validation_steps_s1 > 0:
            fit_kwargs_s1['validation_data'] = validation_generator
            fit_kwargs_s1['validation_steps'] = validation_steps_s1

        history_stage1 = model.fit(train_generator, **fit_kwargs_s1)
        print("\n--- Model Training Stage 1 Complete ---")

    except Exception as e:
        print(f"Error during Stage 1 model training: {e}")
        # Optionally exit or just proceed without Stage 1 results


# --- 8. Model Fine-Tuning (Stage 2 - Full OPGs) ---
print("\n--- Preparing for Fine-Tuning (Stage 2 - Full OPGs) ---")

history_stage2 = None
fine_tuning_performed = False
can_fine_tune = True

if history_stage1 is None:
    print("--- Skipping Fine-Tuning: Stage 1 did not complete successfully. ---")
    can_fine_tune = False
elif not model:
    print("--- Skipping Fine-Tuning: Model not available. ---")
    can_fine_tune = False
elif FINE_TUNE_AT_LAYER <= 0:
    print(f"--- Skipping Fine-Tuning: FINE_TUNE_AT_LAYER ({FINE_TUNE_AT_LAYER}) not positive. ---")
    can_fine_tune = False
elif not validation_generator or validation_steps_s1 == 0:
    print("--- Skipping Fine-Tuning: Validation data missing or empty. ---")
    can_fine_tune = False

if can_fine_tune:
    # Unfreeze layers
    base_model.trainable = True
    print(f"Base model trainable status set to: {base_model.trainable}")
    num_layers_total = len(base_model.layers)
    if 0 < FINE_TUNE_AT_LAYER < num_layers_total:
        print(f"Freezing the first {FINE_TUNE_AT_LAYER} layers of the base model.")
        for layer in base_model.layers[:FINE_TUNE_AT_LAYER]:
            layer.trainable = False
        num_frozen = sum(1 for layer in base_model.layers if not layer.trainable)
        print(f"Number of frozen layers in base model: {num_frozen}")
    else:
        print(f"Warning: FINE_TUNE_AT_LAYER ({FINE_TUNE_AT_LAYER}) invalid or >= num layers ({num_layers_total}). Unfreezing ALL base layers.")

    # Compile with low LR
    model.compile(optimizer=Adam(learning_rate=FINE_TUNE_LEARNING_RATE),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    print("\nModel Summary (Stage 2 - Fine-Tuning):")
    model.summary()

    early_stopping_stage2 = EarlyStopping(
        monitor='val_loss', # Monitor val_loss again
        patience=EARLY_STOP_PATIENCE_STAGE2,
        restore_best_weights=True,
        verbose=1,
        mode='min'
    )

    initial_epoch_ft = 0
    if history_stage1 and hasattr(history_stage1, 'epoch'):
         initial_epoch_ft = len(history_stage1.epoch)
         print(f"Stage 1 ran for {initial_epoch_ft} epochs. Starting fine-tuning epoch numbering after that.")

    total_epochs_combined = EPOCHS

    print("\n--- Starting Model Training (Stage 2: Fine-Tuning) ---")
    try:
        fit_kwargs_s2 = {
            'steps_per_epoch': steps_per_epoch_s1, # Use same steps as stage 1
            'epochs': total_epochs_combined,
            'initial_epoch': initial_epoch_ft,
            'callbacks': [early_stopping_stage2],
            'class_weight': class_weights if class_weights is not None else None,
            'validation_data': validation_generator, # We know it exists here
            'validation_steps': validation_steps_s1, # Use same steps as stage 1
        }
        history_stage2 = model.fit(train_generator, **fit_kwargs_s2)
        print("\n--- Model Training Stage 2 (Fine-Tuning) Complete ---")
        fine_tuning_performed = True

    except Exception as e:
         print(f"Error during Stage 2 (Fine-Tuning) model training: {e}")


# --- 9. Evaluation (on Full OPG Test Set) ---
print("\n--- Evaluating Final Model on Full OPG Test Set ---")

final_history = history_stage2 if fine_tuning_performed and history_stage2 else history_stage1
training_stage_name = "Fine-Tuning (Stage 2)" if final_history == history_stage2 else "Head Training (Stage 1)"

if model is None:
    print("--- Skipping Evaluation: Model not available. ---")
elif final_history is None or not hasattr(final_history, 'history') or not final_history.history:
    print(f"--- Skipping Evaluation: No training history available for stage '{training_stage_name}'. ---")
else:
    print(f"Evaluating model after stage: {training_stage_name}")

    # Plotting history (similar logic as before)
    acc = final_history.history.get('accuracy', [])
    val_acc = final_history.history.get('val_accuracy', [])
    loss = final_history.history.get('loss', [])
    val_loss = final_history.history.get('val_loss', [])
    epochs_run = final_history.epoch
    best_epoch = -1
    best_epoch_idx = -1
    if val_loss:
        best_epoch_idx = np.argmin(val_loss)
        best_epoch = epochs_run[best_epoch_idx]

    plt.figure(figsize=(12, 5))
    plot_created = False
    if acc and val_acc:
        plt.subplot(1, 2, 1)
        plt.plot(epochs_run, acc, label='Training Accuracy')
        plt.plot(epochs_run, val_acc, label='Validation Accuracy')
        if best_epoch != -1 and best_epoch_idx != -1 and val_acc:
             plt.scatter(best_epoch, val_acc[best_epoch_idx], color='red', label=f'Best Val Loss Epoch ({best_epoch})', zorder=5)
        plt.legend(loc='best'); plt.title(f'Full OPG Accuracy ({training_stage_name})'); plt.xlabel('Epoch'); plt.ylabel('Accuracy')
        plot_created = True
    # Add similar logic for plotting only training accuracy if val_acc is missing

    loss_plot_idx = 2 if plot_created else 1
    if loss and val_loss:
        plt.subplot(1, loss_plot_idx + 1, loss_plot_idx)
        plt.plot(epochs_run, loss, label='Training Loss')
        plt.plot(epochs_run, val_loss, label='Validation Loss')
        if best_epoch != -1 and best_epoch_idx != -1 and val_loss:
             plt.scatter(best_epoch, val_loss[best_epoch_idx], color='red', label=f'Best Val Loss Epoch ({best_epoch})', zorder=5)
        plt.legend(loc='best'); plt.title(f'Full OPG Loss ({training_stage_name})'); plt.xlabel('Epoch'); plt.ylabel('Loss')
        plot_created = True
    # Add similar logic for plotting only training loss if val_loss is missing

    if plot_created:
        plt.tight_layout()
        plt.savefig("full_opg_training_history.png")
        print("Saved training history plot to full_opg_training_history.png")
        plt.show()
    else:
        print("No history metrics found to plot.")

    # --- Evaluate on Test Set ---
    if test_generator is None or test_generator.samples == 0:
        print("Warning: No test samples found. Skipping test set evaluation.")
    else:
        print("\nEvaluating on Test Set...")
        test_steps = test_generator.samples
        try:
            test_generator.reset()
            test_loss, test_acc = model.evaluate(test_generator, steps=test_steps, verbose=1)
            print(f"Test Set Loss: {test_loss:.4f}")
            print(f"Test Set Accuracy: {test_acc:.4f}")

            # --- Detailed Metrics ---
            print("\nGenerating Predictions for Detailed Metrics...")
            test_generator.reset()
            Y_pred_probs = model.predict(test_generator, steps=test_steps, verbose=1)
            y_pred_indices = np.argmax(Y_pred_probs, axis=1)
            y_true_indices = test_generator.classes
            if len(y_pred_indices) != len(y_true_indices):
                 print(f"WARNING: Prediction/Label count mismatch! Preds={len(y_pred_indices)}, Labels={len(y_true_indices)}. Truncating.")
                 min_len = min(len(y_pred_indices), len(y_true_indices))
                 y_pred_indices = y_pred_indices[:min_len]
                 y_true_indices = y_true_indices[:min_len]

            print(f"\nClasses used for report/matrix: {generator_class_names}")
            print("\nClassification Report (on Full OPG Test Set):")
            print(classification_report(y_true_indices, y_pred_indices, labels=range(num_classes), target_names=generator_class_names, zero_division=0))

            print("\nConfusion Matrix (on Full OPG Test Set):")
            cm = confusion_matrix(y_true_indices, y_pred_indices, labels=range(num_classes))
            plt.figure(figsize=(10, 8))
            sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=generator_class_names, yticklabels=generator_class_names)
            plt.title('Confusion Matrix (Full OPG Test Set)'); plt.ylabel('True Label'); plt.xlabel('Predicted Label')
            plt.savefig("full_opg_confusion_matrix.png")
            print("Saved confusion matrix plot to full_opg_confusion_matrix.png")
            plt.show()

            # --- ROC Curves ---
            try:
                y_true_bin = label_binarize(y_true_indices, classes=range(num_classes))
                n_classes_roc = y_true_bin.shape[1]
                if n_classes_roc > 1 and Y_pred_probs.shape[1] == n_classes_roc:
                    fpr, tpr, roc_auc = dict(), dict(), dict()
                    plt.figure(figsize=(8, 6))
                    colors = cycle(['aqua', 'darkorange', 'cornflowerblue', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan'])
                    roc_plot_created = False
                    for i, color in zip(range(n_classes_roc), colors):
                        if np.sum(y_true_bin[:, i]) > 0: # Check if class has true samples
                            fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], Y_pred_probs[:, i])
                            roc_auc[i] = auc(fpr[i], tpr[i])
                            plt.plot(fpr[i], tpr[i], color=color, lw=2, label=f'ROC class {generator_class_names[i]} (area = {roc_auc[i]:.2f})')
                            roc_plot_created = True
                    if roc_plot_created:
                        plt.plot([0, 1], [0, 1], 'k--', lw=2)
                        plt.xlim([0.0, 1.0]); plt.ylim([0.0, 1.05])
                        plt.xlabel('False Positive Rate'); plt.ylabel('True Positive Rate')
                        plt.title('Multi-class ROC (Full OPG Test Set)'); plt.legend(loc="lower right")
                        plt.savefig("full_opg_roc_curves.png")
                        print("Saved ROC curve plot to full_opg_roc_curves.png")
                        plt.show()
                    else: print("No ROC curves plotted (no positive samples found?).")
                else: print("Skipping ROC curves (not multi-class or shape mismatch).")
            except Exception as roc_e: print(f"Could not generate ROC curves: {roc_e}")

        except Exception as eval_e:
            print(f"Error during test set evaluation: {eval_e}")


# --- 10. Inference on a Single Full OPG (Revised) ---
def predict_single_opg(image_path, model, target_size=(IMG_HEIGHT, IMG_WIDTH), class_names_list=class_names):
    """Loads, preprocesses, and predicts the class for a single full OPG image."""
    if not os.path.exists(image_path):
        print(f"Error: Image path does not exist: {image_path}")
        return None, None
    if model is None:
         print("Error: Model object is None in predict_single_opg.")
         return None, None

    try:
        img = cv2.imread(image_path)
        if img is None: raise IOError(f"Cannot read image file: {image_path}")

        # Ensure 3 channels (BGR) and resize
        if len(img.shape) == 2 or img.shape[2] == 1: img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
        elif img.shape[2] == 4: img = cv2.cvtColor(img, cv2.COLOR_BGRA2BGR)
        elif img.shape[2] != 3: raise ValueError(f"Unsupported channels: {img.shape[2]}")

        img_resized = cv2.resize(img, target_size, interpolation=cv2.INTER_AREA if img.shape[0]>target_size[1] else cv2.INTER_LINEAR)

        # Rescale like the generator and add batch dimension
        img_rescaled = img_resized / 255.0
        img_batch = np.expand_dims(img_rescaled, axis=0) # Add batch dimension -> (1, H, W, C)

        # Verify shape matches model input
        expected_input_shape = model.input_shape[1:]
        if img_batch.shape[1:] != expected_input_shape:
             raise ValueError(f"Input shape mismatch. Expected {expected_input_shape}, got {img_batch.shape[1:]}")

        # Predict
        predictions = model.predict(img_batch)[0] # Get probabilities for the single image

        # Get predicted class and confidence
        pred_index = np.argmax(predictions)
        pred_class = class_names_list[pred_index]
        pred_confidence = predictions[pred_index]
        all_probs = {cn: p for cn, p in zip(class_names_list, predictions)}


        print(f"\n--- Prediction for Single OPG: {os.path.basename(image_path)} ---")
        print(f"Predicted Class: {pred_class}")
        print(f"Confidence: {pred_confidence:.4f}")
        print(f"All Probabilities: { {k: f'{v:.3f}' for k,v in all_probs.items()} }")
        print("--- End Prediction ---")

        return pred_class, all_probs

    except IOError as e: print(f"I/O Error predicting {image_path}: {e}"); return None, None
    except ValueError as e: print(f"Value Error predicting {image_path}: {e}"); return None, None
    except cv2.error as cv_err: print(f"OpenCV Error predicting {image_path}: {cv_err}"); return None, None
    except Exception as e: print(f"Unexpected Error predicting {image_path}: {e.__class__.__name__}: {e}"); return None, None

# --- Example Usage for Single Prediction ---
print("\n--- Example: Predicting on a Single Full OPG from Test Set ---")

if 'model' in locals() and model is not None and 'original_test_filepaths' in locals() and original_test_filepaths:
    # Select a random original test file path
    try:
        example_img_path = np.random.choice(original_test_filepaths)
        print(f"Using original test image for prediction example: {example_img_path}")
        predicted_class, predicted_probs = predict_single_opg(example_img_path, model)
    except ValueError: # Handle case where original_test_filepaths might be empty
         print("No original test file paths available to select from.")
    except Exception as e:
         print(f"Error selecting or predicting example test image: {e}")

elif 'model' not in locals() or model is None:
     print("Model not available. Skipping single OPG prediction example.")
else: # Model exists but original_test_filepaths list is missing or empty
     print("Original test file list not available or empty. Skipping example prediction based on identified test set images.")
     # Optional: Could try prediction on a raw data file as a fallback here if desired


# --- Optional Cleanup ---
# print(f"\n--- Optional Cleanup: Removing {base_local_dir_full} ---")
# try:
#     if os.path.exists(base_local_dir_full):
#         shutil.rmtree(base_local_dir_full)
#         print(f"Successfully removed {base_local_dir_full}")
#     else:
#         print(f"Directory {base_local_dir_full} not found, skipping cleanup.")
# except Exception as e:
#     print(f"Error during cleanup: {e}")

print("\n--- Script Execution Finished ---")
```

```python
# ******************************************************************************
# ** CRITICAL SETUP STEP FOR SAGEMAKER (If needed) **
# ******************************************************************************
# If you previously encountered the "libGL.so.1" error and haven't fixed it:
# 1. Open a Terminal in SageMaker (File -> New -> Terminal).
# 2. Run: sudo yum install -y mesa-libGL
# 3. Restart the Kernel (Kernel -> Restart Kernel).
# ******************************************************************************

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, Input
# *** CHOOSE A STRONGER BASE MODEL ***
# from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications import ResNet50V2 # Option 1: ResNet50V2
# from tensorflow.keras.applications import EfficientNetB3 # Option 2: EfficientNetB3 (might need tf >= 2.4)
from tensorflow.keras.optimizers import AdamW # Using AdamW often works well
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
import cv2

import sagemaker
import boto3
from urllib.parse import urlparse
import os
import shutil
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.utils import class_weight
from itertools import cycle
import glob
import time

# --- Basic Library Checks ---
print(f"TensorFlow Version: {tf.__version__}")
print(f"SageMaker SDK Version: {sagemaker.__version__}")
try:
    print(f"OpenCV Version: {cv2.__version__}")
except NameError:
    print("ERROR: OpenCV (cv2) could not be imported. Please ensure it and its dependencies (like libGL) are installed.")
    exit(1)

# --- 1. Configuration (Revised for Full OPGs & Improved Model) ---
s3_paths = {
    "Broken Teeth": "s3://sagemaker-studio-010526272250-hsf94lgtf6i/Mendeley_Data/Dental OPG (Classification)/BDC-BDR/",
    "Caries": "s3://sagemaker-studio-010526272250-hsf94lgtf6i/Mendeley_Data/Dental OPG (Classification)/Caries/",
    "Fractured Teeth": "s3://sagemaker-studio-010526272250-hsf94lgtf6i/Mendeley_Data/Dental OPG (Classification)/Fractured Teeth/",
    "Healthy Teeth": "s3://sagemaker-studio-010526272250-hsf94lgtf6i/Mendeley_Data/Dental OPG (Classification)/Healthy Teeth/",
    "Impacted Teeth": "s3://sagemaker-studio-010526272250-hsf94lgtf6i/Mendeley_Data/Dental OPG (Classification)/Impacted teeth/",
    "Infected Teeth": "s3://sagemaker-studio-010526272250-hsf94lgtf6i/Mendeley_Data/Dental OPG (Classification)/Infection/"
}

# --- Make class names filesystem-friendly and sort ---
class_names = sorted([name.replace(" ", "_") for name in s3_paths.keys()])
s3_paths_clean_keys = {name.replace(" ", "_"): path for name, path in s3_paths.items()}
num_classes = len(class_names)

print(f"Classes found: {class_names}")
print(f"Number of classes: {num_classes}")

# --- Local directories (Revised for Full OPGs) ---
base_local_dir_full = '/tmp/dental_data_full_opg_improved' # New base directory
raw_data_dir = os.path.join(base_local_dir_full, 'raw')
processed_data_dir_full = os.path.join(base_local_dir_full, 'processed')
train_dir_full = os.path.join(processed_data_dir_full, 'train')
validation_dir_full = os.path.join(processed_data_dir_full, 'validation')
test_dir_full = os.path.join(processed_data_dir_full, 'test')
model_save_path = os.path.join(base_local_dir_full, 'best_model.keras') # Save best model

# --- Model/Training Parameters (Revised for Full OPGs & Improved Model) ---
# *** INCREASE IMAGE SIZE ***
IMG_HEIGHT = 384 # Try larger size (e.g., 384 or 512)
IMG_WIDTH = 384
IMG_SHAPE = (IMG_HEIGHT, IMG_WIDTH, 3)

# *** ADJUST BATCH SIZE FOR MEMORY ***
BATCH_SIZE = 16 # Reduce batch size if increasing image size causes memory errors
EPOCHS = 75 # Allow more epochs, rely on EarlyStopping & ReduceLROnPlateau
INITIAL_LEARNING_RATE = 5e-4 # Starting LR, will be reduced
FINE_TUNE_LEARNING_RATE = 5e-5 # Lower LR for fine-tuning
WEIGHT_DECAY = 1e-4 # For AdamW

VAL_SPLIT = 0.15
TEST_SPLIT = 0.15
RANDOM_STATE = 42
DROPOUT_RATE = 0.5 # Slightly increase dropout
DENSE_UNITS = 512 # More capacity in the head
# *** Adjust fine-tuning layer based on the chosen model (e.g., ResNet50V2) ***
# Check model.summary() to find a good layer index
# For ResNet50V2, layers[-20] might be a starting point, or fine-tune fewer layers initially
FINE_TUNE_AT_LAYER_NAME = 'conv5_block3_out' # Example: Fine-tune from last block onwards in ResNet50V2
# Find the index corresponding to this name after loading the base model
fine_tune_at_index = -1 # Will be set later

EARLY_STOP_PATIENCE = 15 # Increase patience
REDUCE_LR_PATIENCE = 5
REDUCE_LR_FACTOR = 0.2

# --- 2. Data Download (Same as before - ensure it runs cleanly) ---
# (Copy the download_s3_folder function and the download loop here)
# ... (Assume download happens successfully as in the original code) ...
def download_s3_folder(s3_uri, local_dir):
    """Downloads files from an S3 prefix to a local directory."""
    if not os.path.exists(local_dir):
        os.makedirs(local_dir, exist_ok=True)
        print(f"Created directory: {local_dir}")
    else:
        print(f"Directory {local_dir} already exists. Cleaning...")
        shutil.rmtree(local_dir)
        os.makedirs(local_dir, exist_ok=True)

    print(f"Downloading data from {s3_uri} to {local_dir}...")
    start_time = time.time()
    try:
        sagemaker.s3.S3Downloader.download(s3_uri, local_dir)
        elapsed_time = time.time() - start_time
        if not os.listdir(local_dir):
             print(f"Warning: Download from {s3_uri} completed, but directory {local_dir} is empty.")
        else:
             print(f"Successfully downloaded data from {s3_uri} in {elapsed_time:.2f} seconds")
    except Exception as e:
        print(f"Error downloading {s3_uri}: {e}")
        raise e

print("\n--- Starting Data Download ---")
if os.path.exists(raw_data_dir):
    print(f"Removing existing raw data directory: {raw_data_dir}")
    shutil.rmtree(raw_data_dir)
os.makedirs(raw_data_dir)

download_success = True
for class_name in class_names:
    s3_path = s3_paths_clean_keys.get(class_name)
    if s3_path:
        local_class_dir = os.path.join(raw_data_dir, class_name)
        try:
            download_s3_folder(s3_path, local_class_dir)
        except Exception as e:
            print(f"Download failed for class '{class_name}', stopping execution.")
            download_success = False
            break
    else:
        print(f"Warning: Could not find S3 path for class '{class_name}'.")

if not download_success:
    print("\n--- Data Download Failed. Exiting. ---")
    exit(1)
else:
    print("\n--- Data Download Phase Complete ---")


# --- 3. Data Splitting (Same as before - ensure it runs cleanly) ---
# (Copy the split_original_data function and the splitting call here)
# ... (Assume splitting happens successfully as in the original code) ...
def split_original_data(source_dir, train_dir, validation_dir, test_dir, classes, val_split, test_split, random_state=RANDOM_STATE):
    """Splits original image files."""
    print(f"\n--- Starting Data Splitting for Full OPGs ---")
    for dir_path in [processed_data_dir_full, train_dir, validation_dir, test_dir]:
        if os.path.exists(dir_path):
            print(f"Removing existing directory: {dir_path}")
            shutil.rmtree(dir_path)
    os.makedirs(processed_data_dir_full)
    os.makedirs(train_dir)
    os.makedirs(validation_dir)
    os.makedirs(test_dir)
    for cls in classes:
        os.makedirs(os.path.join(train_dir, cls), exist_ok=True)
        os.makedirs(os.path.join(validation_dir, cls), exist_ok=True)
        os.makedirs(os.path.join(test_dir, cls), exist_ok=True)

    all_original_files = []
    print(f"Scanning source directory: {source_dir}")
    for cls in classes:
        class_source_dir = os.path.join(source_dir, cls)
        if not os.path.isdir(class_source_dir):
             print(f"Warning: Source directory for class '{cls}' not found. Skipping.")
             continue
        supported_extensions = ['*.png', '*.jpg', '*.jpeg', '*.bmp', '*.tif', '*.tiff']
        class_files = []
        for ext in supported_extensions:
            class_files.extend(glob.glob(os.path.join(class_source_dir, ext)))
            class_files.extend(glob.glob(os.path.join(class_source_dir, ext.upper())))
        for f_path in sorted(list(set(class_files))):
            all_original_files.append((f_path, cls))

    if not all_original_files:
        raise ValueError("No source images found for splitting.")
    print(f"Found {len(all_original_files)} total original image files.")

    filepaths = [item[0] for item in all_original_files]
    labels = [item[1] for item in all_original_files]

    try:
        train_val_paths, test_paths, train_val_labels, test_labels = train_test_split(
            filepaths, labels, test_size=test_split, random_state=random_state, stratify=labels)
        relative_val_split = val_split / (1.0 - test_split) if (1.0 - test_split) > 0 else 0
        if relative_val_split >= 1.0 or len(train_val_paths) < 2:
             train_paths, val_paths = train_val_paths, []
             train_labels, val_labels = train_val_labels, []
        else:
             train_paths, val_paths, train_labels, val_labels = train_test_split(
                train_val_paths, train_val_labels, test_size=relative_val_split, random_state=random_state, stratify=train_val_labels)
    except ValueError as e:
        print(f"Warning: Stratified split failed ({e}). Falling back to non-stratified.")
        # Simplified non-stratified fallback
        train_val_paths, test_paths = train_test_split(filepaths, test_size=test_split, random_state=random_state)
        relative_val_split = val_split / (1.0 - test_split) if (1.0 - test_split) > 0 else 0
        if relative_val_split < 1.0 and len(train_val_paths)>0:
             train_paths, val_paths = train_test_split(train_val_paths, test_size=relative_val_split, random_state=random_state)
        else:
             train_paths, val_paths = train_val_paths, []
        # Note: Labels aren't perfectly tracked here, rely on copy counts
        train_labels, val_labels, test_labels = [], [], [] # Dummy labels for copy function

    print(f"Splitting complete: {len(train_paths)} train, {len(val_paths)} validation, {len(test_paths)} test images.")

    def copy_files(file_list, dest_base_dir):
        counts = {cls: 0 for cls in classes}
        for src_path in file_list:
            try:
                filename = os.path.basename(src_path)
                class_name = os.path.basename(os.path.dirname(src_path))
                if class_name in classes:
                    dest_dir = os.path.join(dest_base_dir, class_name)
                    dest_path = os.path.join(dest_dir, filename)
                    shutil.copy2(src_path, dest_path)
                    counts[class_name] += 1
                else:
                     print(f"Warning: Could not determine class for {src_path}. Skipping copy.")
            except Exception as copy_e:
                print(f"Error copying {src_path} to {dest_base_dir}: {copy_e}")
        return counts

    print("Copying training files...")
    train_counts = copy_files(train_paths, train_dir)
    print("Copying validation files...")
    val_counts = copy_files(val_paths, validation_dir)
    print("Copying test files...")
    test_counts = copy_files(test_paths, test_dir)

    print("\n--- File Splitting Summary (Full OPGs) ---")
    total_copied = 0
    for cls in classes:
        tr_c, v_c, te_c = train_counts.get(cls, 0), val_counts.get(cls, 0), test_counts.get(cls, 0)
        print(f"Class '{cls}': Train={tr_c}, Val={v_c}, Test={te_c}")
        total_copied += tr_c + v_c + te_c
    print(f"Total Files Copied: {total_copied}")
    print(f"--- File Splitting Complete ---")
    return test_paths # Return list of test file paths

try:
    original_test_filepaths = split_original_data(raw_data_dir, train_dir_full, validation_dir_full, test_dir_full, class_names, VAL_SPLIT, TEST_SPLIT)
except ValueError as e:
    print(f"Stopping script due to error during data preparation: {e}")
    exit(1)
except Exception as e:
     print(f"An unexpected error occurred during data preparation: {e}")
     exit(1)


# --- 4. Calculate Class Weights (Same logic, but check counts) ---
print("\n--- Calculating Class Weights for Training Set (Full OPGs) ---")
class_weights = None
try:
    train_samples_count = {}
    all_train_labels_indices = []
    class_label_map = {cls: i for i, cls in enumerate(class_names)}
    train_files_found = False

    print(f"Scanning training directory: {train_dir_full}")
    for cls, label_index in class_label_map.items():
        class_path = os.path.join(train_dir_full, cls)
        count = 0
        if os.path.isdir(class_path):
            try:
                files_in_class = [f for f in os.listdir(class_path) if os.path.isfile(os.path.join(class_path, f))]
                count = len(files_in_class)
            except OSError as list_err: count = 0
            if count > 0:
                train_files_found = True
                all_train_labels_indices.extend([label_index] * count)
        train_samples_count[cls] = count
    print(f"Training image counts per class: {train_samples_count}")

    if not train_files_found or not all_train_labels_indices:
         print("Warning: No training samples found. Cannot calculate class weights.")
    else:
        unique_labels = sorted(np.unique(all_train_labels_indices))
        if len(unique_labels) < num_classes: print(f"Warning: Only {len(unique_labels)}/{num_classes} classes in training set.")

        class_weights_array = class_weight.compute_class_weight(
            class_weight='balanced', classes=unique_labels, y=all_train_labels_indices)
        class_weights = {i: 0.0 for i in range(num_classes)}
        for i, label_index in enumerate(unique_labels):
            if label_index < num_classes: class_weights[label_index] = class_weights_array[i]
        print(f"Calculated Class Weights (index: weight): {class_weights}")

except Exception as e:
     print(f"Error calculating class weights: {e}")
     class_weights = None
if class_weights is None: print("Proceeding without class weights.")


# --- 5. Data Augmentation and Generators (Adjusted for larger images) ---
print("\n--- Setting up Data Generators (Improved) ---")

# Consider slightly less aggressive augmentation for larger/detailed images initially
train_datagen = ImageDataGenerator(
    rescale=1./255, # Rescaling is essential
    rotation_range=10,
    width_shift_range=0.05,
    height_shift_range=0.05,
    shear_range=0.05,
    zoom_range=0.05, # Less zoom
    horizontal_flip=True,
    fill_mode='nearest',
    brightness_range=[0.95, 1.05] # Tighter brightness
    )
# Preprocessing function for ResNetV2/EfficientNet (if needed - typically just rescale is fine, but check docs)
# train_datagen = ImageDataGenerator(preprocessing_function=tf.keras.applications.resnet_v2.preprocess_input, ...)

validation_test_datagen = ImageDataGenerator(rescale=1./255)
# validation_test_datagen = ImageDataGenerator(preprocessing_function=tf.keras.applications.resnet_v2.preprocess_input)

train_generator, validation_generator, test_generator = None, None, None
generator_class_names = class_names

try:
    train_generator = train_datagen.flow_from_directory(
        train_dir_full,
        target_size=(IMG_HEIGHT, IMG_WIDTH), # Use updated size
        batch_size=BATCH_SIZE,              # Use updated batch size
        class_mode='categorical',
        classes=class_names,
        color_mode='rgb',
        shuffle=True,
        seed=RANDOM_STATE)

    validation_generator = validation_test_datagen.flow_from_directory(
        validation_dir_full,
        target_size=(IMG_HEIGHT, IMG_WIDTH),
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        classes=class_names,
        color_mode='rgb',
        shuffle=False) # No shuffle for validation

    test_generator = validation_test_datagen.flow_from_directory(
        test_dir_full,
        target_size=(IMG_HEIGHT, IMG_WIDTH),
        batch_size=1, # Keep batch size 1 for test evaluation
        class_mode='categorical',
        classes=class_names,
        color_mode='rgb',
        shuffle=False) # No shuffle for test

    print(f"Found {train_generator.samples} train, {validation_generator.samples} validation, {test_generator.samples} test images.")
    if train_generator.samples == 0: raise RuntimeError("No training images found.")
    print(f"Class Indices: {train_generator.class_indices}")
    generator_class_names = list(train_generator.class_indices.keys())
    if not (train_generator.class_indices == validation_generator.class_indices == test_generator.class_indices):
         raise RuntimeError("Class indices mismatch!")

except Exception as e:
     print(f"Error initializing ImageDataGenerators: {e}")
     # Add checks for directory existence here if needed
     exit(1)

# --- 6. Model Building (Using ResNet50V2 or EfficientNet) ---
print("\n--- Building Model (Using ResNet50V2) ---")
model = None

try:
    # Load the chosen base model
    # Option 1: ResNet50V2
    base_model = ResNet50V2(input_shape=IMG_SHAPE, include_top=False, weights='imagenet')
    # Option 2: EfficientNetB3 (ensure TensorFlow version compatibility)
    # base_model = EfficientNetB3(input_shape=IMG_SHAPE, include_top=False, weights='imagenet')

    # Find the index of the layer to fine-tune from
    try:
        layer_names = [layer.name for layer in base_model.layers]
        fine_tune_at_index = layer_names.index(FINE_TUNE_AT_LAYER_NAME)
        print(f"Will fine-tune from layer '{FINE_TUNE_AT_LAYER_NAME}' (index {fine_tune_at_index}) onwards.")
    except ValueError:
        print(f"Warning: Fine-tune layer name '{FINE_TUNE_AT_LAYER_NAME}' not found in base model '{base_model.name}'. Fine-tuning ALL layers.")
        fine_tune_at_index = 0 # Fine-tune all if name not found

    # Stage 1: Freeze base model initially
    base_model.trainable = False
    print(f"Base model '{base_model.name}' loaded. Trainable (Stage 1): {base_model.trainable}")

    # Build classification head
    inputs = Input(shape=IMG_SHAPE)
    # Apply base model's preprocessing if needed (often handled by generator)
    # x = tf.keras.applications.resnet_v2.preprocess_input(inputs)
    # x = base_model(x, training=False)
    x = base_model(inputs, training=False) # Use training=False when base is frozen
    x = GlobalAveragePooling2D(name="global_avg_pool")(x)
    x = Dropout(DROPOUT_RATE, name="head_dropout")(x) # Increased dropout
    x = Dense(DENSE_UNITS, activation='relu', name="head_dense")(x) # Increased dense units
    outputs = Dense(num_classes, activation='softmax', name="head_softmax")(x)

    model = Model(inputs, outputs)

    # Compile for Stage 1 using AdamW and initial LR
    optimizer_stage1 = AdamW(learning_rate=INITIAL_LEARNING_RATE, weight_decay=WEIGHT_DECAY)
    model.compile(optimizer=optimizer_stage1,
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    print("\nModel Summary (Stage 1 - Head Training):")
    model.summary(line_length=120) # Wider summary

except Exception as e:
    print(f"Error building the model: {e}")
    exit(1)

# --- 7. Model Training (Stage 1 - Improved Callbacks) ---
print("\n--- Starting Model Training (Stage 1: Head Only - Improved) ---")

# Define Callbacks
early_stopping = EarlyStopping(
    monitor='val_loss', # Monitor validation loss
    patience=EARLY_STOP_PATIENCE, # Increased patience
    restore_best_weights=True, # Restore best weights found
    verbose=1,
    mode='min'
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=REDUCE_LR_FACTOR, # Reduce LR significantly
    patience=REDUCE_LR_PATIENCE, # Reduce LR if no improvement for X epochs
    verbose=1,
    min_lr=1e-7 # Don't reduce beyond this
)

# Save the *best* model based on validation loss
model_checkpoint = ModelCheckpoint(
    filepath=model_save_path, # Save to defined path
    monitor='val_loss',
    save_best_only=True,
    save_weights_only=False, # Save entire model
    mode='min',
    verbose=1
)

callbacks_stage1 = [early_stopping, reduce_lr, model_checkpoint]

steps_per_epoch = 0
validation_steps = 0
if train_generator: steps_per_epoch = max(1, train_generator.samples // BATCH_SIZE)
if validation_generator: validation_steps = max(1, validation_generator.samples // BATCH_SIZE)

history_stage1 = None
if not model or steps_per_epoch == 0:
    print("Error: Model not built or no training data. Cannot start Stage 1.")
    exit(1)
elif validation_steps == 0:
    print("Warning: No validation data. Early stopping and LR reduction will monitor 'loss'. Model checkpointing disabled.")
    early_stopping = EarlyStopping(monitor='loss', patience=EARLY_STOP_PATIENCE, restore_best_weights=True, verbose=1, mode='min')
    reduce_lr = ReduceLROnPlateau(monitor='loss', factor=REDUCE_LR_FACTOR, patience=REDUCE_LR_PATIENCE, verbose=1, min_lr=1e-7)
    callbacks_stage1 = [early_stopping, reduce_lr] # Cannot use ModelCheckpoint without val_loss

print(f"Stage 1 Training with: steps_per_epoch={steps_per_epoch}, validation_steps={validation_steps}")
print(f"Using class weights: {class_weights is not None}")
try:
    fit_kwargs_s1 = {
        'steps_per_epoch': steps_per_epoch,
        'epochs': EPOCHS, # Use full epochs, let callbacks handle stopping/LR
        'callbacks': callbacks_stage1,
        'class_weight': class_weights if class_weights is not None else None
    }
    if validation_generator and validation_steps > 0:
        fit_kwargs_s1['validation_data'] = validation_generator
        fit_kwargs_s1['validation_steps'] = validation_steps

    history_stage1 = model.fit(train_generator, **fit_kwargs_s1)
    print("\n--- Model Training Stage 1 Complete ---")

    # Load the best weights saved by ModelCheckpoint if validation was used
    if validation_steps > 0 and os.path.exists(model_save_path):
        print(f"Loading best weights from Stage 1 saved at {model_save_path}")
        model.load_weights(model_save_path) # Load weights back into the current model structure

except Exception as e:
    print(f"Error during Stage 1 model training: {e}")
    # Decide whether to exit or proceed

# --- 8. Model Fine-Tuning (Stage 2 - Improved) ---
print("\n--- Preparing for Fine-Tuning (Stage 2 - Improved) ---")

history_stage2 = None
fine_tuning_performed = False
can_fine_tune = True

if history_stage1 is None:
    print("--- Skipping Fine-Tuning: Stage 1 did not complete successfully. ---")
    can_fine_tune = False
elif not model:
    print("--- Skipping Fine-Tuning: Model not available. ---")
    can_fine_tune = False
elif fine_tune_at_index < 0: # Check if a valid index was found
    print(f"--- Skipping Fine-Tuning: Invalid fine_tune_at_index ({fine_tune_at_index}). ---")
    can_fine_tune = False
elif validation_steps == 0: # Fine-tuning really needs validation data
    print("--- Skipping Fine-Tuning: Validation data missing or empty. Essential for fine-tuning. ---")
    can_fine_tune = False

if can_fine_tune:
    # Unfreeze layers from the specified index
    base_model.trainable = True
    print(f"Base model trainable status set to: {base_model.trainable}")
    print(f"Unfreezing layers from index {fine_tune_at_index} ('{FINE_TUNE_AT_LAYER_NAME}') onwards.")
    for layer in base_model.layers[:fine_tune_at_index]:
        layer.trainable = False

    num_trainable_base = sum(1 for layer in base_model.layers if layer.trainable)
    print(f"Number of trainable layers in base model after unfreezing: {num_trainable_base}")

    # Compile with low LR for fine-tuning
    optimizer_stage2 = AdamW(learning_rate=FINE_TUNE_LEARNING_RATE, weight_decay=WEIGHT_DECAY)
    model.compile(optimizer=optimizer_stage2,
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    print("\nModel Summary (Stage 2 - Fine-Tuning):")
    model.summary(line_length=120)

    # Use similar callbacks, potentially adjusted patience
    early_stopping_ft = EarlyStopping(monitor='val_loss', patience=EARLY_STOP_PATIENCE, restore_best_weights=True, verbose=1, mode='min')
    reduce_lr_ft = ReduceLROnPlateau(monitor='val_loss', factor=REDUCE_LR_FACTOR, patience=REDUCE_LR_PATIENCE, verbose=1, min_lr=1e-8) # Allow slightly lower LR
    # Update checkpoint path or reuse, ensuring it saves the best fine-tuned model
    model_checkpoint_ft = ModelCheckpoint(filepath=model_save_path, monitor='val_loss', save_best_only=True, save_weights_only=False, mode='min', verbose=1)

    callbacks_stage2 = [early_stopping_ft, reduce_lr_ft, model_checkpoint_ft]

    initial_epoch_ft = 0
    if history_stage1 and hasattr(history_stage1, 'epoch'):
        # Find the epoch where Stage 1 stopped (or the best epoch if restore_best_weights was True)
        # Simplest is just number of epochs run
         initial_epoch_ft = len(history_stage1.epoch)
         print(f"Stage 1 ran for {initial_epoch_ft} epochs. Starting fine-tuning after that.")
    else:
         # Estimate based on early stopping if possible, otherwise start from 0
         print("Could not determine Stage 1 epochs accurately, starting fine-tuning epoch count from 0.")

    # Define total epochs for fine-tuning phase (can be same as initial EPOCHS or different)
    fine_tune_epochs = EPOCHS # Let fine-tuning run for up to EPOCHS additional cycles

    print("\n--- Starting Model Training (Stage 2: Fine-Tuning) ---")
    try:
        fit_kwargs_s2 = {
            'steps_per_epoch': steps_per_epoch,
            'epochs': initial_epoch_ft + fine_tune_epochs, # Total epochs from start
            'initial_epoch': initial_epoch_ft, # Start numbering from here
            'callbacks': callbacks_stage2,
            'class_weight': class_weights if class_weights is not None else None,
            'validation_data': validation_generator,
            'validation_steps': validation_steps,
        }
        history_stage2 = model.fit(train_generator, **fit_kwargs_s2)
        print("\n--- Model Training Stage 2 (Fine-Tuning) Complete ---")
        fine_tuning_performed = True

        # Load the best weights saved during fine-tuning
        if os.path.exists(model_save_path):
             print(f"Loading best weights from Fine-Tuning saved at {model_save_path}")
             # Need to load the whole model if saved that way, or just weights
             # If using ModelCheckpoint with save_weights_only=False:
             model = tf.keras.models.load_model(model_save_path)
             # If using save_weights_only=True:
             # model.load_weights(model_save_path)
        else:
             print("Warning: Best fine-tuned model file not found. Using the model state at the end of training.")


    except Exception as e:
         print(f"Error during Stage 2 (Fine-Tuning) model training: {e}")

# --- 9. Evaluation (Using the Best Saved Model) ---
print("\n--- Evaluating Final Model (Best Weights) on Test Set ---")

# Load the best model saved by ModelCheckpoint
final_model = None
if os.path.exists(model_save_path):
    print(f"Loading best overall model from: {model_save_path}")
    try:
        final_model = tf.keras.models.load_model(model_save_path)
        print("Best model loaded successfully.")
    except Exception as e:
        print(f"Error loading saved model: {e}. Evaluating with model state after last training stage.")
        final_model = model # Fallback to the model in memory
else:
    print("Warning: No saved best model found. Evaluating with model state after last training stage.")
    final_model = model # Use the model currently in memory

if final_model is None:
    print("--- Skipping Evaluation: Model not available. ---")
else:
    # Combine histories if fine-tuning occurred
    combined_history = {'accuracy': [], 'val_accuracy': [], 'loss': [], 'val_loss': []}
    start_epoch_s2 = 0
    if history_stage1 and history_stage1.history:
        for key in combined_history.keys():
            combined_history[key].extend(history_stage1.history.get(key, []))
        start_epoch_s2 = len(history_stage1.epoch)

    if history_stage2 and history_stage2.history:
         for key in combined_history.keys():
             combined_history[key].extend(history_stage2.history.get(key, []))

    total_epochs_run = len(combined_history['loss']) # Get total epochs from combined history

    # Plotting combined history
    if total_epochs_run > 0:
        epochs_range = range(total_epochs_run)
        acc = combined_history['accuracy']
        val_acc = combined_history['val_accuracy']
        loss = combined_history['loss']
        val_loss = combined_history['val_loss']

        plt.figure(figsize=(14, 6))

        plt.subplot(1, 2, 1)
        if acc: plt.plot(epochs_range, acc, label='Training Accuracy')
        if val_acc: plt.plot(epochs_range, val_acc, label='Validation Accuracy')
        if fine_tuning_performed: plt.axvline(x=start_epoch_s2 - 0.5, color='r', linestyle='--', label='Fine-Tuning Start')
        plt.legend(loc='best'); plt.title('Combined Training and Validation Accuracy'); plt.xlabel('Epoch'); plt.ylabel('Accuracy')

        plt.subplot(1, 2, 2)
        if loss: plt.plot(epochs_range, loss, label='Training Loss')
        if val_loss: plt.plot(epochs_range, val_loss, label='Validation Loss')
        if fine_tuning_performed: plt.axvline(x=start_epoch_s2 - 0.5, color='r', linestyle='--', label='Fine-Tuning Start')
        # Find best val loss epoch if available
        if val_loss:
             best_epoch_idx = np.argmin(val_loss)
             plt.scatter(best_epoch_idx, val_loss[best_epoch_idx], color='red', zorder=5, label=f'Best Val Loss (Epoch {best_epoch_idx})')
        plt.legend(loc='best'); plt.title('Combined Training and Validation Loss'); plt.xlabel('Epoch'); plt.ylabel('Loss')

        plt.tight_layout()
        plt.savefig("improved_training_history.png")
        print("Saved combined training history plot.")
        plt.show()
    else:
        print("No training history available to plot.")


    # --- Evaluate on Test Set ---
    if test_generator is None or test_generator.samples == 0:
        print("Warning: No test samples found. Skipping test set evaluation.")
    else:
        print("\nEvaluating final model on Test Set...")
        test_steps = test_generator.samples # Evaluate all test samples
        try:
            test_generator.reset() # Ensure generator starts from the beginning
            # Use the loaded best model for evaluation
            test_loss, test_acc = final_model.evaluate(test_generator, steps=test_steps, verbose=1)
            print(f"Test Set Loss (Best Model): {test_loss:.4f}")
            print(f"Test Set Accuracy (Best Model): {test_acc:.4f}")

            # --- Detailed Metrics ---
            print("\nGenerating Predictions for Detailed Metrics (Best Model)...")
            test_generator.reset()
            Y_pred_probs = final_model.predict(test_generator, steps=test_steps, verbose=1)
            y_pred_indices = np.argmax(Y_pred_probs, axis=1)
            y_true_indices = test_generator.classes # Get true labels directly from generator

            # Ensure lengths match (should if steps=test_generator.samples)
            if len(y_pred_indices) != test_generator.samples:
                print(f"Warning: Prediction count ({len(y_pred_indices)}) != Sample count ({test_generator.samples}). Using available predictions.")
                y_true_indices = y_true_indices[:len(y_pred_indices)] # Truncate true labels if needed

            print(f"\nClasses used for report/matrix: {generator_class_names}")
            print("\nClassification Report (on Test Set - Best Model):")
            # Ensure labels cover all possible classes, even if not predicted
            report = classification_report(y_true_indices, y_pred_indices, labels=range(num_classes), target_names=generator_class_names, zero_division=0)
            print(report)

            print("\nConfusion Matrix (on Test Set - Best Model):")
            cm = confusion_matrix(y_true_indices, y_pred_indices, labels=range(num_classes))
            plt.figure(figsize=(10, 8))
            sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=generator_class_names, yticklabels=generator_class_names)
            plt.title('Confusion Matrix (Test Set - Best Model)'); plt.ylabel('True Label'); plt.xlabel('Predicted Label')
            plt.savefig("improved_confusion_matrix.png")
            print("Saved confusion matrix plot.")
            plt.show()

            # --- ROC Curves ---
            # (Keep the ROC curve generation code from the original script)
            try:
                y_true_bin = label_binarize(y_true_indices, classes=range(num_classes))
                n_classes_roc = y_true_bin.shape[1]
                if n_classes_roc > 1 and Y_pred_probs.shape[1] == n_classes_roc:
                    fpr, tpr, roc_auc = dict(), dict(), dict()
                    plt.figure(figsize=(8, 6))
                    colors = cycle(['aqua', 'darkorange', 'cornflowerblue', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan']) # Add more colors if > 6 classes
                    roc_plot_created = False
                    for i, color in zip(range(n_classes_roc), colors):
                         # Check if class actually present in true labels AND has variance
                        if i < len(generator_class_names) and np.sum(y_true_bin[:, i]) > 0 and len(np.unique(y_true_bin[:, i])) > 1:
                            fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], Y_pred_probs[:, i])
                            roc_auc[i] = auc(fpr[i], tpr[i])
                            plt.plot(fpr[i], tpr[i], color=color, lw=2, label=f'ROC {generator_class_names[i]} (AUC = {roc_auc[i]:.2f})')
                            roc_plot_created = True
                        else:
                             print(f"Skipping ROC for class '{generator_class_names[i]}' (index {i}): Not present or no variance in test labels.")

                    if roc_plot_created:
                        plt.plot([0, 1], [0, 1], 'k--', lw=2)
                        plt.xlim([0.0, 1.0]); plt.ylim([0.0, 1.05])
                        plt.xlabel('False Positive Rate'); plt.ylabel('True Positive Rate')
                        plt.title('Multi-class ROC (Test Set - Best Model)'); plt.legend(loc="lower right")
                        plt.savefig("improved_roc_curves.png")
                        print("Saved ROC curve plot.")
                        plt.show()
                    else: print("No valid ROC curves plotted.")
                else: print(f"Skipping ROC curves (n_classes={n_classes_roc}, n_probs={Y_pred_probs.shape[1]}).")
            except Exception as roc_e: print(f"Could not generate ROC curves: {roc_e}")


        except tf.errors.ResourceExhaustedError as mem_err:
             print(f"RESOURCE EXHAUSTED during evaluation: {mem_err}. Try reducing BATCH_SIZE for evaluation or use a smaller model/image size.")
        except Exception as eval_e:
            print(f"Error during test set evaluation: {eval_e}")


# --- 10. Inference on a Single Full OPG (Using the Best Loaded Model) ---
# (Keep the predict_single_opg function, just ensure it uses final_model)
def predict_single_opg(image_path, model_to_use, target_size=(IMG_HEIGHT, IMG_WIDTH), class_names_list=generator_class_names):
    """Loads, preprocesses, and predicts the class for a single full OPG image using the provided model."""
    if not model_to_use:
        print("Error: Model not provided for prediction.")
        return None, None
    if not os.path.exists(image_path):
        print(f"Error: Image path does not exist: {image_path}")
        return None, None

    try:
        img = cv2.imread(image_path)
        if img is None: raise IOError(f"Cannot read image file: {image_path}")

        # Ensure 3 channels (BGR) and resize
        if len(img.shape) == 2 or img.shape[2] == 1: img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
        elif img.shape[2] == 4: img = cv2.cvtColor(img, cv2.COLOR_BGRA2BGR)
        img_resized = cv2.resize(img, target_size, interpolation=cv2.INTER_AREA if img.shape[0]>target_size[1] else cv2.INTER_LINEAR)

        # Rescale like the generator (important!)
        img_rescaled = img_resized / 255.0
        # Preprocessing specific to the model if needed (e.g., for ResNetV2)
        # img_processed = tf.keras.applications.resnet_v2.preprocess_input(img_rescaled)
        # img_batch = np.expand_dims(img_processed, axis=0)
        img_batch = np.expand_dims(img_rescaled, axis=0) # Add batch dimension

        # Predict
        predictions = model_to_use.predict(img_batch)[0]

        pred_index = np.argmax(predictions)
        pred_class = class_names_list[pred_index]
        pred_confidence = predictions[pred_index]
        all_probs = {cn: p for cn, p in zip(class_names_list, predictions)}

        print(f"\n--- Prediction for Single OPG: {os.path.basename(image_path)} ---")
        print(f"Predicted Class: {pred_class}")
        print(f"Confidence: {pred_confidence:.4f}")
        print(f"All Probabilities: { {k: f'{v:.3f}' for k,v in all_probs.items()} }")
        print("--- End Prediction ---")

        return pred_class, all_probs

    except Exception as e:
        print(f"Error predicting single image {image_path}: {e.__class__.__name__}: {e}")
        return None, None

# --- Example Usage for Single Prediction (using the best model) ---
print("\n--- Example: Predicting on a Single Full OPG from Test Set (Best Model) ---")

if final_model and 'original_test_filepaths' in locals() and original_test_filepaths:
    try:
        example_img_path = np.random.choice(original_test_filepaths)
        print(f"Using original test image for prediction example: {example_img_path}")
        # Pass the loaded best model to the prediction function
        predicted_class, predicted_probs = predict_single_opg(example_img_path, final_model)
    except ValueError:
         print("No original test file paths available to select from.")
    except Exception as e:
         print(f"Error selecting or predicting example test image: {e}")
elif not final_model:
     print("Best model not available. Skipping single OPG prediction example.")
else:
     print("Original test file list not available or empty. Skipping prediction.")


# --- Optional Cleanup ---
# print(f"\n--- Optional Cleanup: Removing {base_local_dir_full} ---")
# try:
#     if os.path.exists(base_local_dir_full):
#         shutil.rmtree(base_local_dir_full)
#         print(f"Successfully removed {base_local_dir_full}")
# except Exception as e:
#     print(f"Error during cleanup: {e}")

print("\n--- Improved Script Execution Finished ---")
```
