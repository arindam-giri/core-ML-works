```python
import boto3
import pandas as pd
from io import BytesIO

# Set up S3 client
s3 = boto3.client('s3')
# s3://sagemaker-studio-010526272250-hsf94lgtf6i/test_excel1.xlsx

# S3 bucket and file details
bucket_name = 'sagemaker-studio-010526272250-hsf94lgtf6i'
file_key = 'test_excel1.xlsx'

# Read the file from S3
obj = s3.get_object(Bucket=bucket_name, Key=file_key)
data = obj['Body'].read()3

# Create a BytesIO object
excel_data = BytesIO(data)

# Read the Excel file
df = pd.read_excel(excel_data)

# Print the contents
print(df)

# Optional: Print specific columns or rows
# print(df['Column_Name'])
# print(df.head())
```
